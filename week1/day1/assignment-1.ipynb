```python
# Problem 7: Remove Duplicates from a List

arr = [1, 2, 2, 3, 4, 4, 5]
unique = []

for num in arr:
    if num not in unique:
        unique.append(num)

print(unique)
```

```python
# Problem 8: Second Largest Element
arr = [10, 20, 4, 45, 99]

first = second = -1

for num in arr:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print(second)
```

```python
# Problem 9: Frequency of Characters

s = "aabbc"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
# Q11. Check Prime Number

n = 7
is_prime = True

for i in range(2, n):
    if n % i == 0:
        is_prime = False
        break

if is_prime and n > 1:
    print("Prime")
else:
    print("Not Prime")
```

```python
# Q20. Find GCD (Greatest Common Divisor) of Two Numbers

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

num1 = 48
num2 = 18

result = gcd(num1, num2)
print(f"GCD of {num1} and {num2} is: {result}")
```
