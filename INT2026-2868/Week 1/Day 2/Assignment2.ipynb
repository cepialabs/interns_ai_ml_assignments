```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
#Top 10 most frequent genres

genres = df['listed_in'].dropna().str.split(', ')
genres_exploded = genres.explode()

top_10_genres = genres_exploded.value_counts().head(10)
print("Top 10 Genres:" , top_10_genres)
```

```python
#Movies vs TV Shows distribution

type_distribution = df['type'].value_counts()
print("\nMovies vs TV Shows Distribution:")
print(type_distribution)
```

```python
#Content added per year

content_per_year = df['year_added'].value_counts().sort_index()
print("\nContent Added Per Year:")
print(content_per_year)
```

```python
# Country producing most content

countries = df['country'].dropna().str.split(', ')
countries_exploded = countries.explode()

top_country = countries_exploded.value_counts().head(1)
print("\nTop Content Producing Country:")
print(top_country)
```
