```python
#Average spending per customer

avg_spend_per_customer = (
    df.groupby("customer_id")["amount"]
      .sum()
      .mean()
)
print("Average spend per customer:", avg_spend_per_customer)
```

```python
#Top 10% of spenders
customer_spend = (
    df.groupby("customer_id")["amount"]
      .sum()
)

p90_threshold = customer_spend.quantile(0.9)

top_10_percent_spenders = customer_spend[
    customer_spend >= p90_threshold
]

print("Top 10% spenders:")
print(top_10_percent_spenders)

top_10_percent_df = top_10_percent_spenders.reset_index(name="total_spend")
```

```python
#Skewness in customer engagement metrics 

engagement_skew = skew(df["engagement_metric"].dropna())

print("Engagement metric skewness:", engagement_skew)


if engagement_skew > 1:
    print("Highly right-skewed distribution")
elif engagement_skew > 0.5:
    print("Moderately right-skewed distribution")
elif engagement_skew < -0.5:
    print("Left-skewed distribution")
else:
    print("Approximately symmetric distribution")


import numpy as np
df["log_engagement"] = np.log1p(df["engagement_metric"])
```
