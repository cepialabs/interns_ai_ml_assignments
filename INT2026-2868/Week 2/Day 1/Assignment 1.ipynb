```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
# Load dataset
df = pd.read_csv("Mall_Customers.csv")

# Rename columns for easier handling
df.columns = [
    "customer_id",
    "gender",
    "age",
    "annual_income",
    "spending_score"
]
```

```python
# Average spending per customer

avg_spending = df["annual_income"].mean()

print("Average spending per customer:")
print(avg_spending)
```

```python

# Identify top 10% of spenders

p90_threshold = df["annual_income"].quantile(0.9)

top_10_percent_spenders = df[df["annual_income"] >= p90_threshold]

print("\nTop 10% of spenders:")
print(top_10_percent_spenders)
```

```python
# Detect skewness in engagement metric

engagement_skewness = skew(df["spending_score"])

print("\nSkewness of engagement metric (Spending Score):")
print(engagement_skewness)


if engagement_skewness > 1:
    print("Highly right-skewed engagement distribution")
elif engagement_skewness > 0.5:
    print("Moderately right-skewed engagement distribution")
elif engagement_skewness < -0.5:
    print("Left-skewed engagement distribution")
else:
    print("Approximately symmetric engagement distribution")

# Visualization
plt.hist(df["spending_score"], bins=30)
plt.title("Spending Score Distribution")
plt.xlabel("Spending Score")
plt.ylabel("Number of Customers")
plt.show()
```
