# Handwritten Digit Recognition using MNIST

## 1. Introduction
Handwritten digit recognition is one of the most common applications of machine learning in image classification. It is widely used in applications such as postal code recognition, bank cheque processing, and digitized document analysis.

In this project, a neural network model is developed to recognize handwritten digits from images. The model is trained using the MNIST dataset, which is a well-known dataset in machine learning containing thousands of handwritten digit images.

---

## 2. Objective
The main objective of this project is to build a neural network model that can correctly classify handwritten digits from 0 to 9.

The specific goals include:
- Loading and preprocessing the MNIST dataset
- Building a neural network model
- Training the model using training data
- Evaluating model accuracy
- Visualizing the predictions

---

## 3. Dataset Description
The dataset used in this project is the MNIST Handwritten Digits Dataset.

Dataset characteristics:

- Total images: 70,000
- Training images: 60,000
- Testing images: 10,000
- Image size: 28 × 28 pixels
- Number of classes: 10 (digits 0–9)

Each image is a grayscale image where pixel values range from 0 to 255.

---

## 4. Data Preprocessing
Before training the model, the dataset is preprocessed.

The preprocessing steps include:

1. Loading the dataset using Scikit-learn.
2. Normalizing pixel values by dividing by 255 so that values range between 0 and 1.
3. Splitting the dataset into training and testing sets.

This helps improve the performance and training stability of the model.

---

## 5. Model Development
A Multi Layer Perceptron (MLP) neural network is used for this project.

The model consists of:
- Input layer representing image pixels
- Hidden layers for feature learning
- Output layer for digit classification

The ReLU activation function is used in hidden layers, and the Adam optimizer is used for training.

---

## 6. Model Training
The model is trained using the training dataset. During training, the neural network learns patterns from the images and adjusts its internal weights to correctly classify digits.

The training process iteratively improves the model performance.

---

## 7. Model Evaluation
After training, the model is evaluated using the test dataset.

The following evaluation metrics are used:
- Accuracy Score
- Classification Report
- Confusion Matrix

The model achieves a high accuracy in recognizing handwritten digits.

---

## 8. Visualization
Visualization helps in understanding the model predictions.

The following visualizations are included:
- Displaying sample images from the dataset
- Showing predicted digits
- Plotting a confusion matrix

These visualizations help analyze how well the model performs.

---

## 9. Results
The trained neural network successfully classifies handwritten digits from the MNIST dataset.

The model achieves high prediction accuracy and correctly identifies most digits in the test dataset.

This project demonstrates how neural networks can be applied to image classification tasks. Using the MNIST dataset, the model successfully learns to recognize handwritten digits.

The project provides a good introduction to machine learning concepts such as data preprocessing, model training, evaluation, and visualization.
