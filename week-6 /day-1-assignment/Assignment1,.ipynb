````python
#Simple Neural Network with One Hidden Layer

import numpy as np

# Input dataset
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# Expected output
y = np.array([[0],
              [1],
              [1],
              [0]])

# Sigmoid activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x*(1-x)

# Random weights initialization
np.random.seed(42)
weights_input_hidden = np.random.rand(2,3)
weights_hidden_output = np.random.rand(3,1)

learning_rate = 0.1
epochs = 10000

for i in range(epochs):

    # Forward Propagation
    hidden_input = np.dot(X, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_output)
    predicted_output = sigmoid(final_input)

    # Error
    error = y - predicted_output

    # Backpropagation
    d_output = error * sigmoid_derivative(predicted_output)

    error_hidden = d_output.dot(weights_hidden_output.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights
    weights_hidden_output += hidden_output.T.dot(d_output) * learning_rate
    weights_input_hidden += X.T.dot(d_hidden) * learning_rate

print("Predicted Output:")
print(predicted_output)
````

## How Data Flows from Input to Output During Training

### 1. Input Layer
The input layer receives the dataset.

**Example input:**

```
[0,1]
```

These values represent the features given to the neural network.

---

### 2. Hidden Layer Processing
The inputs are multiplied with weights and passed to the hidden layer.

```
Hidden Input = Input × Weights
```

Then the sigmoid activation function is applied to introduce non-linearity.

```
Hidden Output = sigmoid(Hidden Input)
```

This helps the network learn complex patterns.

---

### 3. Output Layer
The output from the hidden layer is again multiplied with another set of weights.

```
Final Input = Hidden Output × Weights
```

Then the sigmoid function produces the predicted output.

```
Predicted Output = sigmoid(Final Input)
```

---

### 4. Error Calculation
The predicted output is compared with the actual output.

```
Error = Actual Output − Predicted Output
```

This shows how far the prediction is from the correct value.

---

### 5. Backpropagation
The error is sent back through the network.

**Steps:**
- Calculate gradient of the output layer  
- Propagate error to hidden layer  
- Adjust weights to reduce error  

---

### 6. Weight Update
Weights are updated using gradient descent.

```
New Weight = Old Weight + Learning Rate × Gradient
```

This improves the neural network's prediction.

---

### 7. Training Process
The steps **forward propagation → error calculation → backpropagation → weight update** are repeated many times (epochs).

Over time, the model learns the correct pattern in the data.

---

### 8. Result
After training, the neural network predicts outputs close to the expected values.
