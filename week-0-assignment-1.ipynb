```python
#2. Even or Odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#3. Largest of Three Numbers
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

if a >= b and a >= c:
    print("Largest:", a)
elif b >= a and b >= c:
    print("Largest:", b)
else:
    print("Largest:", c)
```

```python
#6. Factorial of a Number
n = int(input("Enter number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

```python
#7. Fibonacci Series
n = int(input("Enter number of terms: "))
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#8. Prime Number Check
num = int(input("Enter number: "))
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
#9. sum of digits
num = int(input("Enter number: "))
sum_digits = 0

while num > 0:
    sum_digits += num % 10
    num //= 10

print("Sum of digits =", sum_digits)
```

```python
#10. Reverse a String
s = input("Enter string: ")
print("Reversed string:", s[::-1])
```

```python
#11. Palindrome Check
s = input("Enter string: ")
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#12. Count Vowels
s = input("Enter string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("Vowel count:", count)
```

```python
#13. tring Length Without len()
s = input("Enter string: ")
count = 0

for _ in s:
    count += 1

print("Length:", count)
```

```python
#18. Word Count in Sentence
sentence = input("Enter sentence: ")
words = sentence.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```

```python
#19. Frequency of Characters (Without Counter)
s = "engineering"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
#20. Armstrong Number Check
num = int(input("Enter number: "))
temp = num
digits = len(str(num))
total = 0

while temp > 0:
    d = temp % 10
    total += d ** digits
    temp //= 10

if total == num:
    print("Armstrong number")
else:
    print("Not Armstrong")
```
