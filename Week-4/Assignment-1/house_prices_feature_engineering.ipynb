```python
import pandas as pd
import numpy as np

from io import StringIO

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
data = """price,area,bedrooms,bathrooms,stories,mainroad,guestroom,basement,hotwaterheating,airconditioning,parking,prefarea,furnishingstatus
13300000,7420,4,2,3,yes,no,no,no,yes,2,yes,furnished
"""

df = pd.read_csv(StringIO(data))
df
```

```python
df['price_per_sqft'] = df['price'] / df['area']
df[['price', 'area', 'price_per_sqft']]
```

```python
categorical_cols = [
    'mainroad', 'guestroom', 'basement',
    'hotwaterheating', 'airconditioning',
    'prefarea', 'furnishingstatus'
]

encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

df
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
```

```python
feature_importance_lr = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lin_model.coef_
})

feature_importance_lr
```

```python
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X, y)
```

```python
feature_importance_tree = pd.DataFrame({
    'Feature': X.columns,
    'Importance': tree_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

feature_importance_tree
```

```python
print("Summary:")
print("1. Created price_per_sqft as a new feature.")
print("2. Encoded categorical variables using Label Encoding.")
print("3. Used Linear Regression and Decision Tree to analyze feature importance.")
print("4. Decision Tree importance shows relative contribution of each feature.")
```
