````python
import pandas as pd
from io import StringIO

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
````

````python
data = """"text","spam"
"Subject: naturally irresistible your corporate identity lt is really hard to recollect a company ... 100 % satisfaction guaranteed ...",1
"""

df = pd.read_csv(StringIO(data))
df
````

````python
X = df['text']
y = df['spam']

vectorizer = CountVectorizer(stop_words='english')
X_vectorized = vectorizer.fit_transform(X)
````

````python
log_model = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
````

````python
log_model = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
````

````python

# Check minimum samples per class
min_class_count = y.value_counts().min()

if min_class_count >= 2:
    n_splits = min(5, min_class_count)
    kfold = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
else:
    kfold = None
    print("Not enough samples per class for cross-validation.")
````

````python
if kfold is not None:
    log_scores = cross_val_score(log_model, X_vectorized, y, cv=kfold, scoring='accuracy')
    rf_scores = cross_val_score(rf_model, X_vectorized, y, cv=kfold, scoring='accuracy')

    print("Logistic Regression CV Accuracy:", log_scores)
    print("Random Forest CV Accuracy:", rf_scores)
````

````python
if kfold is not None:
    print("Average Logistic Regression Accuracy:", log_scores.mean())
    print("Average Random Forest Accuracy:", rf_scores.mean())

    print("Logistic Regression Std Dev:", log_scores.std())
    print("Random Forest Std Dev:", rf_scores.std())
````

# 📧 Spam Email Classification  
## 5-Fold Cross Validation with Random Forest & Logistic Regression

## 📌 Project Overview
This project implements a **Spam Email Classifier** using two supervised machine learning models:

- Logistic Regression
- Random Forest Classifier

The models are evaluated using **5-Fold Cross Validation** to measure performance, stability, and generalization.

---

## 📂 Dataset
The dataset contains:

- `text` → Email content
- `spam` → Label  
  - 0 = Not Spam (Ham)  
  - 1 = Spam  

Text data is converted into numerical features using **Bag of Words (CountVectorizer)**.

---

## 🎯 Objectives
- Convert text data into numerical features
- Train Logistic Regression and Random Forest models
- Apply 5-Fold Cross Validation
- Compare performance metrics
- Analyze model stability and generalization

---

## 🧠 Models Used

### 🔹 Logistic Regression
- Linear classification algorithm
- Works well for high-dimensional sparse text data
- Simple and interpretable

### 🔹 Random Forest
- Ensemble tree-based model
- Handles complex patterns
- More flexible but can overfit on small datasets

---

## 🔁 Cross Validation
- Used **Stratified 5-Fold Cross Validation**
- Ensures each fold maintains class distribution
- Helps reduce overfitting
- Provides better estimate of generalization performance

---

## 📊 Evaluation Metrics
- Accuracy (primary metric)
- Mean Cross-Validation Score
- Standard Deviation (model stability)

---

## 📈 Model Comparison

- Higher **average accuracy** → Better predictive performance  
- Lower **standard deviation** → More stable model  
- Stable CV results → Better generalization  

---

## ⚠️ Important Note
Cross-validation requires:
- Multiple samples
- At least one sample from each class in every fold

Small datasets may require reducing the number of folds.
