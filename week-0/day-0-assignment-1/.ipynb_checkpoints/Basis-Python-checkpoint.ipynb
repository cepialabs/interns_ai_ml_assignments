# 20 Python Basic Programs

A collection of 20 fundamental Python programs.

## 1. Hello World Program

```python
print("Hello, World!")
```

## 2. Simple Arithmetic Operations

```python
a = 10
b = 20
print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {b / a}")
print(f"Modulus: {b % a}")
print(f"Exponentiation: {a ** 2}")
```

## 3. String Manipulation

```python
name = "Python"
text = "Hello, Python Programming"
print(f"Length: {len(name)}")
print(f"Uppercase: {name.upper()}")
print(f"Lowercase: {name.lower()}")
print(f"Replace: {text.replace('Python', 'Java')}")
print(f"Contains 'Pro': {'Pro' in text}")
```

## 4. Variable Assignment and Types

```python
x = 10
y = 20.5
z = "Hello"
flag = True
print(f"Integer: {x}, Type: {type(x)}")
print(f"Float: {y}, Type: {type(y)}")
print(f"String: {z}, Type: {type(z)}")
print(f"Boolean: {flag}, Type: {type(flag)}")
```

## 5. Even or Odd Number

```python
num = 15
if num % 2 == 0:
    print(f"{num} is Even")
else:
    print(f"{num} is Odd")
```

## 6. List Operations

```python
numbers = [1, 2, 3, 4, 5]
print(f"List: {numbers}")
print(f"Length: {len(numbers)}")
print(f"First element: {numbers[0]}")
print(f"Last element: {numbers[-1]}")
numbers.append(6)
print(f"After append: {numbers}")
print(f"Sum: {sum(numbers)}")
```

## 7. For Loop

```python
print("Numbers 1 to 5:")
for i in range(1, 6):
    print(i)
```

## 8. While Loop

```python
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
```

## 9. Function Definition

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))
```

## 10. Dictionary Operations

```python
student = {"name": "John", "age": 20, "grade": "A"}
print(f"Student: {student}")
print(f"Name: {student['name']}")
print(f"Age: {student['age']}")
student['age'] = 21
print(f"Updated: {student}")
```

## 11. Factorial Calculation

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 6: {factorial(6)}")
```

## 12. Find Maximum of Three Numbers

```python
a, b, c = 10, 25, 15
max_num = max(a, b, c)
print(f"Maximum of {a}, {b}, {c} is: {max_num}")
```

## 13. Sum of Natural Numbers

```python
n = 10
sum_n = sum(range(1, n + 1))
print(f"Sum of natural numbers from 1 to {n}: {sum_n}")
```

## 14. Fibonacci Series

```python
n = 7
a, b = 0, 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
print()
```

## 15. Prime Number Check

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
```

## 16. List Comprehension

```python
squares = [x**2 for x in range(1, 6)]
print(f"Squares: {squares}")

even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print(f"Even numbers: {even_numbers}")
```

## 17. String Reversal

```python
text = "Python"
reversed_text = text[::-1]
print(f"Original: {text}")
print(f"Reversed: {reversed_text}")
```

## 18. Count Vowels in a String

```python
text = "Hello World"
vowels = "aeiouAEIOU"
count = sum(1 for char in text if char in vowels)
print(f"Number of vowels in '{text}': {count}")
```

## 19. Find Common Elements in Two Lists

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common = [x for x in list1 if x in list2]
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Common elements: {common}")
```

## 20. Check if String is Palindrome

```python
def is_palindrome(text):
    text = text.lower().replace(" ", "")
    return text == text[::-1]

print(f"Is 'racecar' palindrome? {is_palindrome('racecar')}")
print(f"Is 'hello' palindrome? {is_palindrome('hello')}")
print(f"Is 'A man a plan a canal Panama' palindrome? {is_palindrome('A man a plan a canal Panama')}")
```
