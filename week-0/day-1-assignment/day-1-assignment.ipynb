Program 4 -
Write a Python code to check if a number is prime.


```python
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
```

Program 5 -
Program to Display Message Based on Day of the Week Using Match Case

```python
day = input("""Enter day of the week:
Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday
Your choice: """).strip().lower()

match day:
    case "monday":
        print("Start of the work week")
    case "tuesday":
        print("Second working day")
    case "wednesday":
        print("Midweek day")
    case "thursday":
        print("Almost there")
    case "friday":
        print("Last working day")
    case "saturday":
        print("Weekend")
    case "sunday":
        print("Holiday")
    case _:
        print("Invalid day")
```

Prgram 7 -
Write a Python program to determine if a year is a leap year

```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

print(is_leap_year(2024))
print(is_leap_year(2023))
```

Program 9 -
Write a Python program to calculate the sum of digits in a number.

```python
def sum_of_digits(number):
    return sum(int(digit) for digit in str(number))

print(sum_of_digits(123))
```

Program 10 - Program to Validate an Email Address

```python
email = input("Enter an email address: ")

if email.count("@") == 1:
    at_pos = email.index("@")
    dot_pos = email.rindex(".")
    
    if at_pos > 0 and dot_pos > at_pos + 1 and dot_pos < len(email) - 1:
        print("Valid email address")
    else:
        print("Invalid email address")
else:
    print("Invalid email address")
```

Program 13 - Employee Increment Calculation

```python
salary = float(input("Enter employee salary: "))
rating = input("Enter performance rating (Excellent / Good / Average / Poor): ").strip().lower()

if rating == "excellent":
    increment = salary * 0.25
elif rating == "good":
    increment = salary * 0.15
elif rating == "average":
    increment = salary * 0.08
elif rating == "poor":
    increment = 0
else:
    print("Invalid rating")
    increment = None

if increment is not None:
    print("Increment amount:", increment)
    print("Updated salary:", salary + increment)
```

Program 15 -  Purchase Discount Calculation

 Program 19- Create a dictionary of students and their marks. Write a program to:
Find the topper
Find the average marks

```python
student = {"Nernay": 234, "karunanjali": 355, "Nandini": 346, "kavay": 123}
total = 0
topper = ""
max_marks = 0
for name in student:
    marks = student[name]
    total += marks

    if marks > max_marks:
        max_marks = marks
        topper = name

avg = total / len(student)

print("Topper:", topper)
print("Average marks:", avg)    
```
