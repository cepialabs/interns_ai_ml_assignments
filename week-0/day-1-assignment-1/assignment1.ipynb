Program 7

```python
n = input("Enter a number: ")
#Using math which will remove the leading zeroes
rev = 0
temp = n
while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10
print("Reversed number:", rev)
#It prevails trailing zeroes treat the input as a string not integer
rev = n[::-1]
print("Reversed number:", rev)

```

Program 9

```python
s = input("Enter a string: ")

vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for ch in s:
    if ch.isalpha():  
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

Program 16

```python
lst = list(map(int, input("Enter numbers separated by space: ").split()))
seen = set()
result = []

for x in lst:
    if x not in seen:
        seen.add(x)
        result.append(x)

print("Without duplicates:", result)
```

Program 17

```python
lst = list(map(int, input("Enter numbers separated by space: ").split()))
n = int(input("Enter N: "))

lst_unique = list(set(lst))
lst_unique.sort(reverse=True)

if n <= 0 or n > len(lst_unique):
    print("Invalid N")
else:
    print(f"{n}th largest element is:", lst_unique[n-1])
```

Program 18

```python
email = input("Enter an email address: ")

if email.count("@") == 1:
    at_pos = email.index("@")
    dot_pos = email.rindex(".")
    
    if at_pos > 0 and dot_pos > at_pos + 1 and dot_pos < len(email) - 1:
        print("Valid email address")
    else:
        print("Invalid email address")
else:
    print("Invalid email address")
```

Program 19 - Employee Bonus

```python
salary = float(input("Enter employee salary: "))
rating = input("Enter performance rating (Excellent / Good / Average / Poor): ").strip().lower()

if rating == "excellent":
    bonus = salary * 0.20
elif rating == "good":
    bonus = salary * 0.10
elif rating == "average":
    bonus = salary * 0.05
elif rating == "poor":
    bonus = 0
else:
    print("Invalid rating")
    bonus = None

if bonus is not None:
    print("Bonus amount:", bonus)
    print("Total payout:", salary + bonus)
```
