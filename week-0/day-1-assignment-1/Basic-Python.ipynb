```python
# Problme 13:- Given a dictionary of students and marks, find the student with the highest marks.

students = {
    "Rahul": 78,
    "Aman": 92,
    "Neha": 85,
    "Priya": 90
}

top_student = None
highest_marks = -1

for name, marks in students.items():
    if marks > highest_marks:
        highest_marks = marks
        top_student = name

print(top_student, highest_marks)
```

```python
# Problem 14:- Given a dictionary, swap its keys and values.
students = {
    "Rahul": 78,
    "Aman": 92,
    "Neha": 85,
}
new_dict = {}
for key, value in students.items():
    new_dict[value] = key

print(new_dict)
```

```python
# Problem 15:- Write a program to find the factorial of a number using a function and loop.
num1 = int(input("Enter the number"))

factorial = 1

for i in range(1, num1 + 1):
    factorial *= i
print(f"The factorial of {num1} is {factorial}")
```

```python
# Problem 17:-Create a function "calculate" in which it is having add functionality, subtraction functionality, multiplication and division functionality of two numbers.
def calculate(val1, val2, operation = 'add'):
  if operation == 'add':
    return val1 + val2
  elif operation == 'subtract':
    return val1 - val2
  elif operation == 'multiply':
    return val1 * val2
  elif operation == 'divide':
    return val1 / val2
result = calculate(10, 20, "multiply")
result
```

```python
# Problem 18:- Sort the above dictionary in the ascending order using lambda function
learner_scores = {
    'Siddharth': 88,
    'Prem': 89,
    'Mohit':93,
    'Swati':90
}
sorted_values = dict(sorted(learner_scores.items(), key = lambda item : item[1]))
print(sorted_values)
```
