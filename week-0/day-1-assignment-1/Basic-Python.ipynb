8) Check Leap Year

```python
year = int(input("Enter year: "))
if(year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print('Leap year')
else:
    print('Not Leap year')
```

9) Find Factorial


```python
n = int(input('Enter a Number: '))
fact = 1
for i in range(1, n+1):
    fact *= i 
print('Factorial: ', fact)
```

10) Fibonacci Series

```python
n = int(input('Enter a number: '))
a, b = 0, 1
for i in range(n):
    print(a, end=' ')
    a, b = b, a + b
```

11) Check Prime Number

```python
n = int(input("Enter number: "))
for i in range(2, n):
    if n % i == 0:
        print('Not prime')
        break
else:
    print('Prime')
```

17) Check Armstrong Number

```python
n = int(input('Enter a number: '))
temp = n 
sum = 0
while temp > 0:
    d = temp % 10
    sum += d ** 3
    temp //= 10
print('Armstrong ' if sum == n else 'Not Armstrong')
```
