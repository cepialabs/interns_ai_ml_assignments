# 06- program to check whether a given number is prime or not.

```python
num = int(input("Enter a number:"))
if (num<=1):
    print("Its not a prime number!")
elif(num==2):
    print("Its a prime number!")
else:
    for i in range(2,num):
        if(num%i==0):
            print("Its not a prime number!")
            break
        else:
            print("Its a prime number!")
            break
    
```

# 07- Write a program to reverse a given integer number.

```python
n = int(input("Enter a number: "))
rev = 0
while n>0:
    rev = rev * 10 + n % 10
    n //= 10
print(f"The reverse of the number is",rev)
```

# 10- Program to print the first N numbers of the Fibonacci series.

```python
n = int(input("Enter N: "))
a = 0
b = 1
for i in range(n):
    print(a,end="")
    c = a+b
    a = b
    b = c
```

# 18- Create a dictionary of students and their marks. Write a program to:

#####        Find the topper

#####        Find the average marks

```python
student = {"Krushna": 234, "Bhaba": 355, "Binay": 346, "Prasanta": 123}
total = 0
topper = ""
max_marks = 0
for name in student:
    marks = student[name]
    total += marks

    if marks > max_marks:
        max_marks = marks
        topper = name

avg = total / len(student)

print("Topper:", topper)
print("Average marks:", avg)    
```

# 20- Program to check whether a number is an Armstrong number.

```python
n = int(input("Enter a number: "))
temp = n
power = len(str(n))
s = 0

while temp > 0:
    digit = temp % 10
    s += digit ** power
    temp //= 10

if s == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```
