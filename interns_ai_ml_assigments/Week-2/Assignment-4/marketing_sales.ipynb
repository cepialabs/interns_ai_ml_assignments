```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("marketing_sales.csv")

df.head()
```

```python
df.info()

df.describe()
```

```python
correlation_matrix = df[['TV', 'Radio', 'Social Media', 'Sales']].corr()

correlation_matrix
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix: Ad Spend vs Sales")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title("TV Ad Spend vs Sales")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Radio', y='Sales', data=df)
plt.title("Radio Ad Spend vs Sales")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Social Media', y='Sales', data=df)
plt.title("Social Media Spend vs Sales")
plt.show()
```
