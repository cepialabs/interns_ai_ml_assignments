```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("marketing_campaign.csv", sep=';')
df.head()
```

```python
spend_columns = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]

df['Total_Spending'] = df[spend_columns].sum(axis=1)
df[['ID', 'Total_Spending']].head()
```

```python
average_spending = df['Total_Spending'].mean()
print("Average Spending per Customer:", round(average_spending, 2))
```

```python
top_10_threshold = df['Total_Spending'].quantile(0.90)

top_10_spenders = df[df['Total_Spending'] >= top_10_threshold]

print("Top 10% Spending Threshold:", top_10_threshold)
print("Number of Top 10% Spenders:", top_10_spenders.shape[0])
```

```python
engagement_columns = [
    'NumWebVisitsMonth',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumDealsPurchases'
]

skewness_values = df[engagement_columns].skew()
skewness_values
```

```python
for col, value in skewness_values.items():
    if value > 0:
        print(f"{col}: Positively skewed")
    elif value < 0:
        print(f"{col}: Negatively skewed")
    else:
        print(f"{col}: Symmetric")
```
