# Student Performance Dataset Analysis

This notebook performs data visualization on a Student Performance Dataset.

## Import Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Create Dataset

```python

np.random.seed(42)
n = 200

data = pd.DataFrame({
    "Gender": np.random.choice(["Male", "Female"], n),
    "Attendance": np.random.randint(50, 100, n),
    "Math_Score": np.random.randint(35, 100, n),
    "Science_Score": np.random.randint(30, 100, n),
    "English_Score": np.random.randint(40, 100, n)
})

data["Average_Score"] = data[["Math_Score", "Science_Score", "English_Score"]].mean(axis=1)
data.head()
```

## Score Distribution Plot

```python

plt.figure()
plt.hist(data["Average_Score"])
plt.title("Score Distribution")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

## Gender vs Performance

```python

gender_perf = data.groupby("Gender")["Average_Score"].mean()

plt.figure()
plt.bar(gender_perf.index, gender_perf.values)
plt.title("Gender vs Average Performance")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

## Correlation Heatmap

```python

corr = data[["Attendance", "Math_Score", "Science_Score", "English_Score", "Average_Score"]].corr()

plt.figure()
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
```

## Attendance vs Score Trend

```python

plt.figure()
plt.scatter(data["Attendance"], data["Average_Score"])
plt.title("Attendance vs Average Score")
plt.xlabel("Attendance (%)")
plt.ylabel("Average Score")
plt.show()
```
