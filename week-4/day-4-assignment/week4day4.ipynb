```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_validate
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate

```

```python
df = pd.read_csv(
    r"C:\Users\Seerat Metkari\Downloads\emails.csv"
)

# Remove all Unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

df.head()
```

```python
df.columns
```

```python
df = df.dropna()
df = df.drop_duplicates()
df['text'] = df['text'].str.lower()
import re

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['text'] = df['text'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

X = vectorizer.fit_transform(df['text'])
y = df['spam']
```

```python
log_model = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
scoring = ['accuracy', 'precision', 'recall', 'f1']
```

```python
log_5cv = cross_validate(log_model, X, y, cv=5, scoring=scoring)
rf_5cv = cross_validate(rf_model, X, y, cv=5, scoring=scoring)


print("Logistic Regression - 5 Fold")
for metric in scoring:
    print(metric, ":", log_5cv['test_' + metric].mean())

print("\nRandom Forest - 5 Fold")
for metric in scoring:
    print(metric, ":", rf_5cv['test_' + metric].mean())
```
