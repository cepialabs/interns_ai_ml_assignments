# House Prices Prediction using Cross-Validation
Compare Logistic Regression and Random Forest using 5-fold and 10-fold CV.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('house_prices_dataset.csv')
X = data.drop('SalePrice', axis=1)
y = data['SalePrice']
```

```python
# Initialize models
lr = LinearRegression()
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
# 5-Fold CV
lr_scores_5 = cross_val_score(lr, X, y, cv=5, scoring='r2')
rf_scores_5 = cross_val_score(rf, X, y, cv=5, scoring='r2')

print('Linear Regression 5-Fold R2:', lr_scores_5.mean())
print('Random Forest 5-Fold R2:', rf_scores_5.mean())
```

```python
# 10-Fold CV
lr_scores_10 = cross_val_score(lr, X, y, cv=10, scoring='r2')
rf_scores_10 = cross_val_score(rf, X, y, cv=10, scoring='r2')

print('Linear Regression 10-Fold R2:', lr_scores_10.mean())
print('Random Forest 10-Fold R2:', rf_scores_10.mean())
```
