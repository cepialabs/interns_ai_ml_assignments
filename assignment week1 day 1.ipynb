```python
# Calculate average price


house_prices = np.random.randint(1000000, 10000000, size=10000)

average_price = np.mean(house_prices)

print("Average House Price:", average_price)
```

```python
# Remove outliers

# Generate data
house_prices = np.random.randint(1000000, 10000000, size=10000)

# Calculate mean and std
mean = np.mean(house_prices)
std = np.std(house_prices)


filtered_prices = house_prices[(house_prices > mean - 2*std) & 
                                (house_prices < mean + 2*std)]

print("Original count:", len(house_prices))
print("After removing outliers:", len(filtered_prices))



```

```python
import numpy as np

# Generate house prices
house_prices = np.random.randint(1000000, 10000000, size=10000)

# Calculate quartiles
Q1 = np.percentile(house_prices, 25)
Q3 = np.percentile(house_prices, 75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
clean_prices = house_prices[(house_prices >= lower_bound) & 
                             (house_prices <= upper_bound)]

print("Original count:", len(house_prices))
print("After removing outliers:", len(clean_prices))
```
