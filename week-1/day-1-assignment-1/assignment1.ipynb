```python
import numpy as np
import matplotlib.pyplot as plt
```

## Fetching random 10000 prices of the house

```python
prices = np.random.normal(loc=300000, scale=100000, size=10000)
```

## Calculating Average

```python
prices = np.abs(prices)
average_price = np.mean(prices)
print(f"Average Price: ${average_price:,.2f}")
```

## Removing Outliers

```python
Q1 = np.percentile(prices, 25)
Q3 = np.percentile(prices, 75)
IQR = Q3 - Q1
# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
clean_prices = prices[(prices >= lower_bound) & (prices <= upper_bound)]
```

```python
print(f"Original count: {len(prices)}")

print(f"Count after removing outliers: {len(clean_prices)}")

print(f"Outliers removed: {len(prices) - len(clean_prices)}")
```

## Visualization

```python
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=50, alpha=0.5, label='Original Data', color='blue', edgecolor='black')
plt.hist(clean_prices, bins=50, alpha=0.5, label='Cleaned Data', color='green', edgecolor='black')
plt.axvline(average_price, color='red', linestyle='dashed', linewidth=2, label=f'Avg: ${average_price:,.0f}')
plt.title('House Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
