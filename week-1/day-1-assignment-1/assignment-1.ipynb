# Generate 10,000 random house prices using NumPy: 
#### 1. Calculate the average price
#### 2. Remove outliers
#### 3. Visualize price distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Generating the data

```python
np.random.seed(42)

data= np.random.normal(loc=5000000, scale=1000000, size=10000)
```

```python
data
```

# 1. Calculate the average price

```python
avg_price = np.mean(data)
```

```python
float(avg_price)
```

# 3(1). Visualize price distribution before removing outlier.

```python
plt.hist(data, color="green", edgecolor="black")
plt.title("House Price Distribution (Before Removing Outliers)")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

# 2. Remove outliers

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Keeping only the valid prices
clean_prices = data[(data >= lower_bound) & (data <= upper_bound)]
```

```python
clean_prices
```

# 3(2). Visualize price distribution after removing outlier.

```python
plt.hist(clean_prices, color="cyan", edgecolor="black")
plt.title("House Price Distribution (Before Removing Outliers)")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
new_price = clean_prices.mean()
```
