```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/student_performance_dataset.csv")
```

```python
dataset.head()
```

# Score Distribution Plot

```python
scores = ['math_score', 'reading_score', 'writing_score']

dataset[scores].hist(bins=15, figsize=(10,6))
plt.suptitle("Score Distribution of Students")
plt.show()
```

# Gender vs Performance chart

```python
gender_perf = dataset.groupby('gender')[scores].mean()

gender_perf.plot(kind='bar', figsize=(8,5))
plt.title("Gender vs Average Performance")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

# Correlation Heatmap

```python
corr = dataset[['attendance_percentage', 'math_score',
           'reading_score', 'writing_score']].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Attendance vs score

```python
plt.figure(figsize=(6,4))
plt.scatter(dataset['attendance_percentage'], dataset['math_score'])
plt.xlabel("Attendance Percentage")
plt.ylabel("Math Score")
plt.title("Attendance vs Math Score")
plt.show()
```
