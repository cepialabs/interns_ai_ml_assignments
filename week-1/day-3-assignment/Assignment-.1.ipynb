```python
#Handle Missing Age Values
print("Missing Age before:", df['Age'].isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].median())

print("Missing Age after:", df['Age'].isnull().sum())
```

```python
#Remove Duplicate Rows
print("Duplicates before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates after:", df.duplicated().sum())
```

```python
#Fix Inconsistent Gender Formats
print("Before fixing gender:")
print(df['Gender'].value_counts())

df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].replace({'m':'male','f':'female'})

print("\nAfter fixing gender:")
print(df['Gender'].value_counts())
```

```python
#Salary Outliers
df.columns = df.columns.str.strip()

# Outlier removal using IQR on Total Spend
col = 'Total Spend'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Before removing outliers:")
print(df[col].describe())

df = df[(df[col] >= lower) & (df[col] <= upper)]

print("\nAfter removing outliers:")
print(df[col].describe())
```

```python
#Encode
df.columns = df.columns.str.strip()

# Encode categorical columns
df = pd.get_dummies(
    df,
    columns=['Gender', 'Subscription Type', 'Contract Length'],
    drop_first=True
)

print("Columns after encoding:\n", df.columns)
```

```python
#Feature Engineering
df['SpendTenureRatio'] = df['Total Spend'] / (df['Tenure'] + 1)

df[['SpendTenureRatio']].head()
```
