```python
#Remove duplicates
df1 = df.drop_duplicates()
df1
```

```python
#Handle Missing Age Values
print("Missing Age before:", df['Age'].isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].median())

print("Missing Age after:", df['Age'].isnull().sum())
```

```python
#Remove Duplicate Rows
print("Duplicates before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates after:", df.duplicated().sum())
```

```python
#Fix Inconsistent Gender Formats
print("Before fixing gender:")
print(df['Gender'].value_counts())

df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].replace({'m':'male','f':'female'})

print("\nAfter fixing gender:")
print(df['Gender'].value_counts())
```

```python
#Salary Outliers
print("Salary before cleaning:", df['EstimatedSalary'].describe())

Q1 = df['EstimatedSalary'].quantile(0.25)
Q3 = df['EstimatedSalary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['EstimatedSalary'] >= lower) & (df['EstimatedSalary'] <= upper)]

print("\nSalary after removing outliers:")
print(df['EstimatedSalary'].describe())
```

```python
#Encode
df = pd.get_dummies(df, columns=['Geography','Gender'], drop_first=True)
print("Columns after encoding:\n", df.columns)
```

```python
#Scale
num_cols = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

for col in num_cols:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

df[num_cols].head()
```

```python
#Feature Engineering
df['BalanceSalaryRatio'] = df['Balance'] / (df['EstimatedSalary'] + 1)
df[['BalanceSalaryRatio']].head()
```
