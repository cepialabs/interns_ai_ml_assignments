```python
#Clean column names
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
print(df.columns)
```

```python
#Rename correctly
df.rename(columns={
    "product_titl": "product_title",
    "total_rating": "review_count",
    "next_month_offer": "offer_text",
    "product_de": "product_details_dict",
    "model_nam": "model_name"
}, inplace=True)
print(df.columns.tolist())
```

```python
#Clean Price, MRP, Reviews
for col in ["price", "mrp", "review_count"]:
    df[col] = (df[col].astype(str)
               .str.replace("₹", "", regex=False)
               .str.replace(",", "", regex=False)
               .str.replace("+", "", regex=False)
               .str.replace(" bought", "", regex=False)
               .str.strip())
    df[col] = pd.to_numeric(df[col], errors="coerce")
df[["price","mrp","review_count"]].head()
```

```python
#Extract Brand
df["brand_clean"] = df["product_details_dict"].astype(str)\
    .str.extract(r"'Brand':\s*'([^']+)'")

df[["product_details_dict", "brand_clean"]].head()
```
