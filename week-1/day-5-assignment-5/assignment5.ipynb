## 1.Load and understand the dataset

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("EDA analysis.csv")
df.head()
df.shape
df.info()
df.describe()
```

## 2. Clean the dataset

```python
df.isnull().sum().sort_values(ascending=False)
# categorical
df["Alley"].fillna("None", inplace=True)
df["PoolQC"].fillna("None", inplace=True)

# numerical
df["LotFrontage"].fillna(df["LotFrontage"].median(), inplace=True)
df
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df = df[df["SalePrice"] < 500000]
sns.boxplot(x=df["SalePrice"])
```

## 3.Perform meaningful analysis

```python
df.corr(numeric_only=True)["SalePrice"].sort_values(ascending=False)  #which features affect the price of the house
```

## 4.Create at least SIX visualizations

```python
sns.histplot(df["SalePrice"], kde=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
```

```python
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)
```

```python
sns.barplot(x="OverallQual", y="SalePrice", data=df)
```

```python
sns.lineplot(x="YearBuilt", y="SalePrice", data=df)
```

```python
df.groupby("Neighborhood")["SalePrice"].mean().sort_values().plot(kind="bar")
```
