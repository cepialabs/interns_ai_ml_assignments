```python
GENERATE 10,000 RANDOM HOUSE PRICES USING NUMPY AND CALUCULATE
1. CALCULATE AVERAGE PRICE
2. REMOVE OUTLIERS
3. VISUALIZE PRICES DISTRIBUTION
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

#Generate 10,000 random house prices
prices=np.random.randint(10_00_000, 1_00_00_000, 10000)

#Calculate average price
average_price=np.mean(prices)
print("Average House Price:",average_price)
```

```python
#Remove outliers using IQR
Q1 = np.percentile(prices, 25)
Q3 = np.percentile(prices, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_prices = prices[(prices >= lower_bound) & (prices <= upper_bound)]
print("Prices count after removing outliers:", len(filtered_prices))
```

```python
#Visualize price distribution 
plt.hist(filtered_prices, bins=50)
plt.title("House Price Distribution (Without Outliers)")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#line plot
plt.scatter(range(len(filtered_prices)),filtered_prices)
plt.title("House Prices Scatter plot")
plt.xlabel("Index")
plt.ylabel("House Price")
plt.show()
```
