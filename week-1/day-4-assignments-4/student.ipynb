```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("student-mat.csv")

# View first 5 rows
df.head()
```

```python
df.info()
```

### Score Distribution plot

```python
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10)
plt.xlabel("Final Score (G3)")
plt.ylabel("Number of Students")
plt.title("Score Distribution of Students")
plt.show()
```

### Gender VS Performance Chart

```python
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='G3', data=df)
plt.xlabel("Gender")
plt.ylabel("Average Final Score (G3)")
plt.title("Gender vs Performance")
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(10,8))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

### Attendance Vs Score Trend Chart

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='absences', y='G3', data=df)
plt.xlabel("Number of Absences")
plt.ylabel("Final Score (G3)")
plt.title("Attendance vs Final Score")
plt.show()
```
