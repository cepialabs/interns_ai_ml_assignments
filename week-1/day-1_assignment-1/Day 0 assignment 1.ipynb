```python
#Check Whether a Number is Abundant Number or not
num = int(input("Enter a number : "))
length = len(str(num))
sqr = str(num ** 2)
print(sqr)

new = int(sqr[(length):])
print(new)

if (new == num):
    print("It is an Automorphic Number")
else:
    print("It is not an Automorphic Number")
```

```python
# Print the Armstrong Numbers in Nth Range
range = int(input("Enter the range : "))

for i in range(1, range):
    add = 0 
    length = len(str(i))
    temp = i 
    while (temp != 0):
        reminder = temp % 10 
        add = add + (reminder**length)
        temp = temp // 10 
    
    if (i==add):
        print(i)
```

```python
# Check Whether a Number is Palindrome or Not
num = int(input("Enter a number: "))

temp = num
new = 0 
while (temp!=0):
    reminder = temp % 10 
    new = (new * 10) + reminder
    temp = temp // 10
    
print(new)

if (num == new):
    print("The Number is Palindrome")
else:
    print("The Number is not a palindrome")
```

```python
#Print the Reverse of a Number
num = int(input("Enter a number: "))

temp = num
new = 0 
while (temp!=0):
    reminder = temp % 10 
    new = (new * 10) + reminder
    temp = temp // 10
    
print(new)
```

```python
# Greatest Of Three Numbers
n1, n2, n3 = map(int, input("Enter Three Numbers : ").split())

if (n1>=n2) and (n1>=n3):
    print(f'{n1} is the greatest')
elif (n2>=n1) and (n2>=n3):
    print(f'{n2} is the greatest')
elif (n3>=n1) and (n3>=n2):
    print(f'{n3} is the greatest')
else:
    print("They are all Equal")
```

```python
# Checking Whether a number is prime or not
num = int(input("Enter a number : "))

flag = 0
if (num>1):
    for i in range(2, num):
        if (num%i==0):
            flag = 1 
            break
    if flag == 1:
        print("Composite Number")
    else:
        print("Prime Number")
    
else:
    print("Neither Prime Nor Composite")
```

```python
#Check whether a number is Perfect Number or Not
num = int(input("Enter a number : "))
add = 0

for i in range(1, num):
    if num % i == 0:
        add = add + i

if (num == add):
    print("Perfect Number")
else:
    print("Not Perfect Number")
```

```python
#Check Whether or Not the Number is a Strong Number
n = int(input("Enter a number : "))

digits = []
for i in str(n):
    digits.append(int(i))
add = 0  
for i in digits:
    product = 1
    for j in range(1, i+1):
        product = product * j
    add = add + product
print(add)
if (n==add):
    print("It is a strong number")
else:
    print("It is not a strong number")
```
