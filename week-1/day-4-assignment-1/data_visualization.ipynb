```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
np.random.seed(42)

df = pd.DataFrame({
    "Year": range(2015, 2025),
    "Sales": np.random.randint(200, 500, 10),
    "Profit": np.random.randint(50, 200, 10),
    "Expenses": np.random.randint(100, 300, 10)
})

df
```

```python
# Line Plot
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Sales"], marker='o', label="Sales")
plt.plot(df["Year"], df["Profit"], marker='o', label="Profit")
plt.title("Sales & Profit Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

```python
# Bar Chart
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Year"], df["Expenses"])
plt.title("Yearly Expenses")
plt.xlabel("Year")
plt.ylabel("Expenses")
plt.show()
```

```python
# Histogram
```

```python
plt.figure(figsize=(7,5))
plt.hist(df["Sales"], bins=5)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Sales"])
plt.title("Box Plot of Sales (Outliers)")
plt.show()
```

```python
# Heatmap
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
```

```python
sns.pairplot(df)
plt.show()
```

```python
# Styling Plots
```

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(7,5))
sns.lineplot(x="Year", y="Sales", data=df)
plt.title("Styled Line Plot using Seaborn")
plt.show()
```

```python
# Choosing the Right Chart
```

```python
 Line chart  trends over time  
 Bar chart  comparison between categories  
 Histogram  data distribution  
 Box plot  outliers and spread  
 Heatmap  correlation analysis  
 Pairplot  multivariate relationships
```
