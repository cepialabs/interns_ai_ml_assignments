# **Pefrom an analysis on the generated house_prices data. Calculate and display the average price, then identify and remove outliers. And Visualize the distribution of house prices both before and after outlier removal using a histogram.**

```python
import numpy as np

np.random.seed(42)

# Generate 10,000 random house prices (in INR)
house_prices = np.random.randint(100, 12000, size=10_000)

house_prices
```

```python
house_prices.shape
```

## Calculate Average Price

### Subtask:
Calculate and display the average price of the generated house prices.


**Reasoning**:
To calculate the average price of the `house_prices` array, I will use the `.mean()` method and then print the result with a descriptive label as requested.



```python
average_price = house_prices.mean()
print(f"Average House Price: {average_price:,.2f} INR")
```

## Remove Outliers

### Subtask:
Identify and remove outliers from the house prices data using the Interquartile Range (IQR) method.


**Reasoning**:
To identify and remove outliers, I will calculate the first quartile (Q1), third quartile (Q3), Interquartile Range (IQR), and then determine the lower and upper bounds. Finally, I will filter the `house_prices` array based on these bounds to create a new array without outliers.



```python
Q1 = np.percentile(house_prices, 25)
Q3 = np.percentile(house_prices, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

house_prices_filtered = house_prices[(house_prices >= lower_bound) & (house_prices <= upper_bound)]

print(f"Q1: {Q1:,.2f} INR")
print(f"Q3: {Q3:,.2f} INR")
print(f"IQR: {IQR:,.2f} INR")
print(f"Lower Bound for Outliers: {lower_bound:,.2f} INR")
print(f"Upper Bound for Outliers: {upper_bound:,.2f} INR")
print(f"Original number of house prices: {len(house_prices)}")
print(f"Number of house prices after outlier removal: {len(house_prices_filtered)}")
```

**Reasoning**:
The next step is to visualize the distribution of house prices both before and after outlier removal using a histogram. Since no outliers were removed in the previous step, the two distributions are expected to be identical, but it's important to visualize this as part of the task.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

# Histogram for original house prices
plt.subplot(1, 2, 1)
sns.histplot(house_prices, kde=True, color='skyblue')
plt.title('Distribution of Original House Prices')
plt.xlabel('House Price (INR)')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x') # Prevent scientific notation on x-axis

# Histogram for filtered house prices
plt.subplot(1, 2, 2)
sns.histplot(house_prices_filtered, kde=True, color='lightcoral')
plt.title('Distribution of House Prices After Outlier Removal')
plt.xlabel('House Price (INR)')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x') # Prevent scientific notation on x-axis

plt.tight_layout()
plt.show()

print("Histograms showing the distribution of house prices before and after outlier removal have been generated.")
```
