```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Analysing the dataset

```python
data = pd.read_csv("netflix_titles.csv")
```

```python
text = ['director', 'cast', 'country', 'rating', 'duration']
for col in text:
    data[col] = data[col].fillna("Unknown")
data['date_added'] = data['date_added'].fillna(data['date_added'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

# 1. Identify the top 10 most frequent genres.

```python
genres = data['listed_in'].str.split(', ')
all_genres = genres.explode()

top10_genres = all_genres.value_counts().head(10)
```

```python
top10_genres
```

# Top 10 most frequent genres in bar plot.

```python
plt.figure(figsize=(10,5))
plt.bar(top10_genres.index, top10_genres.values, color="orange", edgecolor="black")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

# 2. Compare Movies vs TV Shows distribution

```python
distribution = data['type'].value_counts()
```

```python
distribution
```

# Movies vs TV Shows distribution using a bar chart

```python
plt.bar(distribution.index, distribution.values, color = "cyan", edgecolor = "black") 
plt.title("Movies vs TV Shows Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

# Movies vs TV Shows distribution using the pie chart

```python
plt.pie(distribution.values,labels=distribution.index, autopct="%1.1f%%") 
plt.title("Movies vs TV Shows Distribution")
plt.show()
```

# 3. Analyze content added per year

```python
data['year_added'] = data['date_added'].dt.year
content = data['year_added'].value_counts().sort_index()
```

```python
content
```

# Analyze content added per year in bar chart

```python
plt.figure(figsize=(12,6))
plt.bar(content.index, content.values, color="magenta", edgecolor = "black")
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

# Analyze content added per year in line chart

```python
plt.figure(figsize=(12,6))
plt.plot(content.index, content.values, marker="o")
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

# 4. Identify the country producing the most content

```python
countries = data['country'].str.split(', ')
all_countries = countries.explode()
top_countries = all_countries.value_counts()
```

```python
top_countries
```

```python
top10_countries = top_countries.head(10)
```

```python
top10_countries
```

```python
plt.figure(figsize=(12,6))
plt.bar(top10_countries.index, top10_countries.values, color="magenta", edgecolor="black")
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```
