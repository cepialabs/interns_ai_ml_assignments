# 1. Score distribution plot

```python
plt.figure(figsize=(8,5))
plt.hist(df["math_score"], bins=10, edgecolor = "black")
plt.title("Math Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

# 2. Gender vs Performance Chart

```python
gender_avg = df.groupby("gender")["math_score"].mean()
```

```python
gender_avg
```

```python
plt.figure(figsize=(6,5))
plt.bar(gender_avg.index, gender_avg.values, edgecolor = "black")
plt.title("Gender vs Average Math Score")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

# 3. Correlation Heatmap

```python
corr = df[["math_score", "science_score", "english_score", "attendance"]].corr()
```

```python
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# 4. Attendance vs score trend chart

```python
plt.figure(figsize=(8,5))
plt.scatter(df["attendance"], df["math_score"])
plt.title("Attendance vs Math Score")
plt.xlabel("Attendance (%)")
plt.ylabel("Math Score")
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[["attendance", "math_score"]])
plt.show()
```
