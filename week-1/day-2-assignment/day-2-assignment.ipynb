```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
#Task 1: Top 10 Most Frequent Genres
genres = df['listed_in'].dropna().str.split(', ')
genres_exploded = genres.explode()

top10_genres = genres_exploded.value_counts().head(10)
print(top10_genres)
```

```python
#plot
plt.figure(figsize=(10,5))
top10_genres.plot(kind='bar')
plt.title("Top 10 Most Frequent Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 2: Compare Movies vs TV Shows Distribution
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
#Plot
plt.figure(figsize=(6,4))
type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Movies vs TV Shows Distribution")
plt.ylabel("")
plt.show()
```

```python
#Task 3: Analyze Content Added Per Year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

content_per_year = df['year_added'].value_counts().sort_index()
print(content_per_year)
```

```python
#Plot
plt.figure(figsize=(12,5))
content_per_year.plot(kind='line', marker='o')
plt.title("Netflix Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()
```

```python
#Task 4: Country Producing Most Content
countries = df['country'].dropna().str.split(', ')
countries_exploded = countries.explode()

top_countries = countries_exploded.value_counts().head(10)
print(top_countries)
```

```python
#Plot
plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
