```python
#Remove duplicates 
```

```python
df=df.drop_duplicates()
df
```

```python
genres = df['listed_in'].str.split(', ')
genres_exploded = genres.explode()

top_genres = genres_exploded.value_counts().head(10)
top_genres
```

```python
#TASK 1:Identify Top 10 Most Frequent Genres
```

```python
top_genres.plot(kind='bar', title='Top 10 Genres on Netflix')
plt.show()
```

```python
#TASK 2:Compare Movies vs TV Shows Distribution
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Movies vs TV Shows'
)
plt.ylabel('')
plt.show()
```

```python
#TASK 3: Analyze Content Added Per Year
```

```python
import pandas as pd

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df_year = df.dropna(subset=['year_added'])
content_per_year = df_year['year_added'].value_counts().sort_index()
content_per_year
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
content_per_year.plot(kind='bar')
plt.title('Content Added Per Year on Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
#TASK 4:Identify Country Producing Most Content
```

```python
top_countries = df['country'].str.split(', ').explode().value_counts().head(10)
top_countries
```

```python
top_countries.plot(kind='bar', title='Top 10 Content Producing Countries')
plt.show()
```
