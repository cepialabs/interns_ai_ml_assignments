```python
#NETFLIX Titles Data Analysis

#libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix.csv")
df.head()
df.shape
df.info()
```

```python
#1: Top 10 Most Frequent Genres
genres = df['listed_in'].str.split(', ').explode()

top_10_genres = genres.value_counts().head(10)
top_10_genres
```

```python
#Visualization
top_10_genres.plot(kind='bar')
plt.title("Top 10 Most Frequent Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
#2: Movies vs TV Shows Distribution
type_distribution = df['type'].value_counts()
type_distribution
```

```python
#Visualization
type_distribution.plot(kind='bar')
plt.title("Movies vs TV Shows Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
#3: Content Added Per Year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

content_per_year = df['year_added'].value_counts().sort_index()
content_per_year
```

```python
#Visualization
content_per_year.plot(kind='line')
plt.title("Netflix Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
#4: Country Producing Most Content
countries = df['country'].str.split(', ').explode()

top_countries = countries.value_counts().head(10)
top_countries
```

```python
#Visualization
top_countries.plot(kind='bar')
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```
