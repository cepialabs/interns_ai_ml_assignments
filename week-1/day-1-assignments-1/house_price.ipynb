```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate 10,000 random house prices
# We'll use a mean of $500k and some spread, then scale it
np.random.seed(42)  # For reproducibility
prices = np.random.lognormal(mean=13, sigma=0.5, size=10000)

# Calculate initial average
initial_avg = np.mean(prices)
print(f"Initial Average Price: ${initial_avg:,.2f}")

# 2. Remove Outliers using the IQR Method
Q1 = np.percentile(prices, 25)
Q3 = np.percentile(prices, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data
filtered_prices = prices[(prices >= lower_bound) & (prices <= upper_bound)]
new_avg = np.mean(filtered_prices)

print(f"New Average Price (after removing outliers): ${new_avg:,.2f}")
print(f"Number of outliers removed: {len(prices) - len(filtered_prices)}")

# 3. Visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(filtered_prices, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of House Prices (Outliers Removed)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
