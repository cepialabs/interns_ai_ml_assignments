```python
import pandas as pd
import matplotlib.pyplot as plt
#load data
df = pd.read_csv(r"C:\Users\Malli Mounika\Downloads\student_data.csv")
df
```

```python
df.head()
```

```python
df.describe()
```

```python
#Create Final Score Column
df["score"] = df["G3"]
print("Score column created successfully")
```

```python
#1. Score Distribution Plot
plt.figure()
plt.hist(df["score"], bins=10)
plt.xlabel("Final Score (G3)")
plt.ylabel("Number of Students")
plt.title("Score Distribution")
plt.show()
```

```python
#2. Gender vs Performance Chart
gender_avg = df.groupby("sex")["score"].mean()

plt.figure()
plt.bar(gender_avg.index, gender_avg.values)
plt.xlabel("Gender")
plt.ylabel("Average Final Score")
plt.title("Gender vs Performance")
plt.show()
```

```python
#3. Correlation Heatmap
corr = df[["G1", "G2", "G3", "absences"]].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#4. Attendance vs Score Trend Chart
plt.figure()
plt.scatter(df["absences"], df["score"])
plt.xlabel("Absences")
plt.ylabel("Final Score (G3)")
plt.title("Absences vs Final Score")
plt.show()
```

```python
#4a). Line Trend (sorted by absences)
df_sorted = df.sort_values("absences")

plt.figure()
plt.plot(df_sorted["absences"], df_sorted["score"])
plt.xlabel("Absences")
plt.ylabel("Final Score (G3)")
plt.title("Attendance (Absences) vs Score Trend")
plt.show()
```

```python
#4b).Average Score per Absence Group
avg_by_abs = df.groupby("absences")["score"].mean()

plt.figure()
plt.plot(avg_by_abs.index, avg_by_abs.values, marker='o')
plt.xlabel("Absences")
plt.ylabel("Average Final Score")
plt.title("Average Score vs Absences")
plt.show()
```
