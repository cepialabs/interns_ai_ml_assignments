```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
df = pd.read_csv("student_performance_dataset.csv")

print("Shape:", df.shape)
print(df.head())
print(df.columns)
```

```python
#Basic Cleaning
# Check missing values
print("Missing Values:\n", df.isnull().sum())
# Remove duplicates if any
df = df.drop_duplicates()
# Convert sex values to proper names (optional)
df["Gender"] = df["sex"].map({"M": "Male", "F": "Female"})
# Use G3 as Final Score
df["FinalScore"] = df["G3"]
```

```python
#Create Attendance (%) Column
max_absences = df["absences"].max()

df["Attendance"] = 100 - (df["absences"] / max_absences) * 100
df["Attendance"] = df["Attendance"].round(2)

print(df[["absences", "Attendance"]].head())
```

```python
#TASK 1: Score Distribution Plot (Histogram)
#Shows how students final scores are distributed.
plt.figure(figsize=(10,5))
sns.histplot(df["FinalScore"], bins=15, kde=True)
plt.title("Final Score Distribution (G3)", fontsize=14)
plt.xlabel("Final Score", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#TASK 2: Gender vs Performance Chart (Bar Plot)
#Compare Male vs Female average score
plt.figure(figsize=(7,5))
sns.barplot(x="Gender", y="FinalScore", data=df, errorbar=None)
plt.title("Gender vs Performance (Average Final Score)", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Average Final Score", fontsize=12)
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```

```python
#TASK 3: Correlation Heatmap
#Shows relationships between numeric columns like:
#studytime
#absences
#attendance
#scores
plt.figure(figsize=(12,7))
# Select numeric columns
numeric_df = df.select_dtypes(include=np.number)
corr = numeric_df.corr()
sns.heatmap(corr, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Student Performance Dataset)", fontsize=14)
plt.show()
```

```python
#TASK 4: Attendance vs Score Trend Chart
#Relationship between attendance and final score
#Scatter plot + regression line
plt.figure(figsize=(10,5))
sns.regplot(x="Attendance", y="FinalScore", data=df, scatter_kws={"alpha":0.6})
plt.title("Attendance vs Final Score Trend", fontsize=14)
plt.xlabel("Attendance (%)", fontsize=12)
plt.ylabel("Final Score (G3)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```
