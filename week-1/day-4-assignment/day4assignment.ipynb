```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(6,4))

plt.hist(df['math score'], bins=10, alpha=0.5, label='Math')
plt.hist(df['reading score'], bins=10, alpha=0.5, label='Reading')
plt.hist(df['writing score'], bins=10, alpha=0.5, label='Writing')

plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Math, Reading, and Writing Scores")
plt.legend()
plt.show()
```

```python
gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
gender_avg


```

```python
gender_avg.plot(kind='bar', figsize=(7,5))
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Gender vs Student Performance")
plt.show()
```

```python
score_cols = ['math score', 'reading score', 'writing score', 'attendance']
corr = df[score_cols].corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['attendance'], df['math score'], alpha=0.5, label='Math')
plt.scatter(df['attendance'], df['reading score'], alpha=0.5, label='Reading')
plt.scatter(df['attendance'], df['writing score'], alpha=0.5, label='Writing')
plt.xlabel("Attendance (%)")
plt.ylabel("Score")
plt.title("Attendance vs Scores")
plt.legend()
plt.show()
```
