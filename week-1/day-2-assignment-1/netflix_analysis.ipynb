```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
# Top 10 Most Frequent Genres
```

```python
genres = df['listed_in'].str.split(', ')
genres_exploded = genres.explode()

top_genres = genres_exploded.value_counts().head(10)
top_genres
```

```python
top_genres.plot(kind='bar', title='Top 10 Genres on Netflix')
plt.show()
```

```python
# Movies vs TV Shows Distribution
```

```python
df['type'].value_counts()
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Movies vs TV Shows')
plt.ylabel('')
plt.show()
```

```python
# Content Added Per Year
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

content_per_year = df['year_added'].value_counts().sort_index()
content_per_year.tail(10)
content_per_year.plot(kind='line', title='Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Country Producing Most Content
```

```python
countries = df['country'].dropna().str.split(', ')
countries_exploded = countries.explode()

countries_exploded.value_counts().head(10)
countries_exploded.value_counts().head(10).plot(kind='bar', title='Top 10 Content Producing Countries')
plt.show()
```
