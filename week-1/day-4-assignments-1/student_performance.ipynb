1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Styling
sns.set(style="whitegrid")
```

2. Load the Dataset

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

3. Create Average Score (Important for Analysis)

```python
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df.head()
```

Task 1: Score Distribution Plot

```python
plt.figure(figsize=(8,5))
sns.histplot(df['average_score'], bins=15, kde=True)
plt.title("Distribution of Average Student Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

 Interpretation:
Shows how overall student performance is distributed across the dataset.

Task 2: Gender vs Performance Chart

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='gender', y='average_score', data=df)
plt.title("Gender vs Student Performance")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

 Interpretation:
Compares score spread and median performance between male and female students.

Task 3: Correlation Heatmap

```python
plt.figure(figsize=(7,5))
score_df = df[['math score', 'reading score', 'writing score']]
sns.heatmap(score_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Exam Scores")
plt.show()
```

 Interpretation:
Reading and writing scores show strong positive correlation, while math has moderate correlation.
