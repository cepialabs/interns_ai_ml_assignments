```python
#Housing Price Prediction
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\Malli Mounika\Downloads\Housing1.csv")
data
```

```python
#Convert yes/no to 1/0

yes_no_cols = ['mainroad','guestroom','basement',
               'hotwaterheating','airconditioning','prefarea']

for col in yes_no_cols:
    data[col] = data[col].map({'yes':1, 'no':0})
```

```python
#Convert furnishingstatus


data['furnishingstatus'] = data['furnishingstatus'].map({
    'unfurnished':0,
    'semi-furnished':1,
    'furnished':2
})
```

```python
#Separate X and y

X = data.drop('price', axis=1).values
y = data['price'].values
```

```python
# Normalize features
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
#Add bias column

X = np.hstack((np.ones((X.shape[0],1)), X))
```

```python
#Gradient Descent

theta = np.zeros(X.shape[1])
alpha = 0.01
epochs = 2000

for i in range(epochs):
    y_pred = X.dot(theta)
    error = y_pred - y
    theta = theta - (alpha/len(y)) * X.T.dot(error)
print(theta)
```

```python
#Prediction function
mean = X[:,1:].mean(axis=0)
std = X[:,1:].std(axis=0)

def predict_house(features):
    features = (features - mean) / std
    features = np.insert(features, 0, 1)
    result = features.dot(theta)
    print("Predicted Price:", result)
    return result

predict_house(np.array([7420,4,2,3,1,0,1,0,1,2,1,2]))
```
