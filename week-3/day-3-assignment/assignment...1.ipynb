```python
#Split Features and Label

X = df.drop(columns=['Label']).values
y = df['Label'].values

print(X.shape, y.shape)
```

```python
#Train–Test Split (Manual)

np.random.seed(42)

indices = np.random.permutation(len(X))
split = int(0.8 * len(X))

X_train = X[indices[:split]]
X_test = X[indices[split:]]

y_train = y[indices[:split]]
y_test = y[indices[split:]]
```

```python
#standardization
mean = X_train.mean(axis=0)
std = X_train.std(axis=0) + 1e-8

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
#KNN FUNCTION

def knn_predict(X_train, y_train, X_test, k=5):
    predictions = []

    for test_point in X_test:
        distances = np.sqrt(np.sum((X_train - test_point) ** 2, axis=1))
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices].astype(int)
        predictions.append(np.bincount(k_labels).argmax())

    return np.array(predictions)
```

```python
def knn_predict(X_train, y_train, X_test, k=5):
    predictions = []

    for x in X_test:
        distances = np.linalg.norm(X_train - x, axis=1)
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]

        #  FIX: convert (-1, 1) → (0, 1)
        k_labels = np.where(k_labels == -1, 0, 1)

        prediction = np.bincount(k_labels).argmax()
        predictions.append(prediction)

    return np.array(predictions)
```

```python
y_test = np.where(y_test == -1, 0, 1)
```

```python
#KNN Accuracy
y_pred_knn = knn_predict(X_train, y_train, X_test, k=5)
knn_accuracy = np.mean(y_pred_knn == y_test)
print("KNN Accuracy:", knn_accuracy)
```

```python
#ENTROPY

def entropy(y):
    values, counts = np.unique(y, return_counts=True)
    probs = counts / counts.sum()
    return -np.sum(probs * np.log2(probs))
```

```python
#INFORMATION GAIN

def information_gain(X, y, feature_index, threshold):
    left = y[X[:, feature_index] <= threshold]
    right = y[X[:, feature_index] > threshold]

    if len(left) == 0 or len(right) == 0:
        return 0

    total_entropy = entropy(y)
    weighted_entropy = (len(left)/len(y))*entropy(left) + (len(right)/len(y))*entropy(right)

    return total_entropy - weighted_entropy
```

```python
#BEST SPLIT

def best_split(X, y):
    best_gain = 0
    split_feature = None
    split_threshold = None

    for feature in range(X.shape[1]):
        thresholds = np.unique(X[:, feature])[:5]
        for t in thresholds:
            gain = information_gain(X, y, feature, t)
            if gain > best_gain:
                best_gain = gain
                split_feature = feature
                split_threshold = t

    return split_feature, split_threshold
```

```python
#TRAIN SIMPLE DECISION TREE (STUMP)

def train_tree(X, y):
    feature, threshold = best_split(X, y)

    if feature is None:
        return np.bincount(y).argmax()

    return feature, threshold
```
