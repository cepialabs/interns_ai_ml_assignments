# KNN & Decision Tree Classification

Spam Email Classification and House Price Prediction

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.datasets import fetch_california_housing
```

## KNN – Spam Email Classification

```python

# Load spam dataset
df = pd.read_csv("spam_email_dataset.csv")

X = df[['money_word_count', 'has_offer', 'has_link']]
y = df['is_spam']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred))
```

## Decision Tree – Spam Email Classification

```python

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

## Decision Tree – House Price Prediction

```python

housing = fetch_california_housing()
X_h = housing.data
y_h = housing.target

X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(X_h, y_h, test_size=0.25, random_state=42)

dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(X_train_h, y_train_h)

score = dt_reg.score(X_test_h, y_test_h)
print("Decision Tree Regression R^2 Score:", score)
```
