## 1. Load and Understand the Dataset

```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("week2_day1_eda_dataset.csv")
df.head()
```

## 2. Data Cleaning

```

# Check missing values
df.isnull().sum()
```

No missing values found in the dataset.

## 3. Meaningful Analysis & Visualizations

### 1️⃣ Age Distribution

```

plt.hist(df['age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### 2️⃣ Attendance Distribution

```

sns.histplot(df['attendance'], kde=True)
plt.title("Attendance Distribution")
plt.show()
```

### 3️⃣ Gender vs Average Score

```

sns.boxplot(x='gender', y='average_score', data=df)
plt.title("Gender vs Average Score")
plt.show()
```

### 4️⃣ Correlation Heatmap

```

plt.figure(figsize=(8,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### 5️⃣ Study Hours vs Performance

```

plt.scatter(df['study_hours'], df['average_score'])
plt.xlabel("Study Hours")
plt.ylabel("Average Score")
plt.title("Study Hours vs Performance")
plt.show()
```

### 6️⃣ Attendance vs Performance

```

plt.scatter(df['attendance'], df['average_score'])
plt.xlabel("Attendance")
plt.ylabel("Average Score")
plt.title("Attendance vs Performance")
plt.show()
```


## 4. Insights

1. Students with higher attendance generally score better.  
2. Study hours show a positive correlation with average score.  
3. Gender-wise performance difference is minimal.  
4. Strong correlation exists among subject scores.  
5. Extremely low attendance often results in lower performance.
