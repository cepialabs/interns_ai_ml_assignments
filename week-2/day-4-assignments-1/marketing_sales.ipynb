1. Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Load Dataset

```python
df = pd.read_csv("marketing_sales.csv")
df.head()
```

3. Dataset Information

```python
df.info()
```

4. Descriptive Statistics

```python
df.describe()
```

5. Check Missing Values

```python
df.isnull().sum()
```

6. Correlation Matrix

```python
numeric_df = df[['TV','Radio','Social Media','Sales']]
corr_matrix = numeric_df.corr()
corr_matrix
```

7. Heatmap Visualization

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Advertising Spend and Sales")
plt.show()
```

8. Correlation Interpretation

Correlation Interpretation

TV advertising shows an extremely strong positive correlation with Sales (~0.99).

Radio advertising shows a strong positive correlation (~0.87).

Social Media shows a moderate positive correlation (~0.53).

This indicates that increased advertising spend is associated with increased sales. However, correlation only measures association and does not prove causation.

9. Correlation vs Causation

Correlation vs Causation

Correlation describes how two variables move together, while causation means one variable directly affects another.

Although advertising spend and sales are strongly correlated, this does not guarantee that advertising alone caused the sales increase. Establishing causality requires controlled experiments, A/B testing, or regression models with control variables.

Key principle: Correlation â‰  Causation

10. Possible Confounders

Possible Confounders

Confounders are external factors influencing both ad spend and sales:

Influencer category

Seasonal demand

Brand awareness

Discounts and promotions

Product quality

Competitor advertising

Economic conditions

These variables may distort the true relationship between advertising and sales.

11. Final Conclusion

Conclusion

The analysis shows a strong positive correlation between advertising channels and sales, with TV having the strongest relationship. However, correlation alone cannot establish causality. Sales are also affected by influencer type, seasonality, promotions, and market conditions. Therefore, while advertising contributes to sales growth, it is not the sole determinant. Causal conclusions require more advanced modeling or experimental approaches.
