```python
import pandas as pd
from scipy.stats import skew
```

```python
df = pd.read_csv("Marketing campaign dataset.csv")
df.head()
```

```python
df.info()
```

```python
#Calculate Average Spending per Customer
avg_spending = df.groupby("advertiser_id")["media_cost_usd"].mean()
avg_spending.head()
```

```python
#Identify Top 10% of Spenders
total_spend = df.groupby("advertiser_id")["media_cost_usd"].sum()
threshold = total_spend.quantile(0.90)
top_10_spenders = total_spend[total_spend >= threshold]
top_10_spenders.head()
```

```python
#Detect Skewness in Customer Engagement
engagement_skewness = skew(df["clicks"])
engagement_skewness
```
