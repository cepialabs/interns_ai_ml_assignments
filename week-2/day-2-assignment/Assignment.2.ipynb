```python
#Email Marketing Campaign – Conditional Probability Analysis

#Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Malli Mounika\Downloads\filtered_dataset.csv")
df
```

```python
df = pd.DataFrame(df)
df
```

```python
#Basic Understanding
df.info()
```

```python
df.describe()
```

```python
#Probability that a user clicks an email
df['Clicked'] = df['click_date'].notna().astype(int)
```

```python
total_users = len(df)
clicked_users = df['Clicked'].sum()

p_click = clicked_users / total_users

print("Total users:", total_users)
print("Users who clicked:", clicked_users)
print("Probability of click:", round(p_click, 3))
```

```python
df['Clicked'].value_counts().plot(kind='bar')
plt.xlabel("Clicked (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Probability of Email Click")
plt.show()
```

```python
#Probability that a user converts given they clicked
df['Clicked'] = df['click_date'].notna().astype(int)
df['Converted'] = df['transaction_date'].notna().astype(int)
```

```python
total_users = len(df)
clicked_users = df['Clicked'].sum()

p_click = clicked_users / total_users
print("Probability that user clicks:", round(p_click, 3))
```

```python
clicked_df = df[df['Clicked'] == 1]
converted_after_click = clicked_df['Converted'].sum()

p_convert_given_click = converted_after_click / len(clicked_df)
print("P(Convert | Click):", round(p_convert_given_click, 3))
```

```python
#Conditional Probability Table
df['Clicked'] = df['click_date'].notna().astype(int)
df['Converted'] = df['transaction_date'].notna().astype(int)
```

```python
import pandas as pd

cond_table = pd.crosstab(df['Clicked'], df['Converted'], normalize='index')
cond_table
```

```python
#Joint Probability Table
joint_table = pd.crosstab(df['Clicked'], df['Converted'], normalize=True)
joint_table
```

```python
#Visualization – Click Distribution
df['Clicked'].value_counts().plot(kind='bar')
plt.xlabel("Clicked (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Email Click Distribution")
plt.show()
```

```python
#Visualization – Conversion after Click
clicked_df['Converted'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Conversion Given Click")
plt.ylabel("")
plt.show()
```

```python
#Final Results (Print Nicely)
print("Total Users:", total_users)
print("Users Clicked:", clicked_users)
print("P(Click):", round(p_click, 2))
print("P(Convert | Click):", round(p_convert_given_click, 2))
```
