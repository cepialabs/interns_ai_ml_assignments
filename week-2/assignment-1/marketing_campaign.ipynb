```python
import pandas as pd

# load data
df = pd.read_csv("marketing_campaign.csv")

# average spending per customer
avg_spend = df.groupby("customer_id")["spend"].sum().mean()

avg_spend
```

```python
# total spend per customer
customer_spend = df.groupby("customer_id")["spend"].sum()

# 90th percentile threshold
threshold = customer_spend.quantile(0.90)

# top 10% spenders
top_10_percent = customer_spend[customer_spend >= threshold]

top_10_percent.sort_values(ascending=False)
```

```python
engagement_cols = ["clicks", "email_opens", "website_visits"]

skewness = df[engagement_cols].skew()

skewness
```
