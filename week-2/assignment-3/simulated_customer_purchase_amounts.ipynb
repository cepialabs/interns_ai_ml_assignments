```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# For reproducibility
np.random.seed(42)

# Simulate customer purchase amounts (example dataset)
purchase_amounts = np.random.normal(loc=75, scale=20, size=1000)

# Remove invalid (negative) purchase values
purchase_amounts = purchase_amounts[purchase_amounts > 0]
```

```python
plt.figure()
plt.hist(purchase_amounts, bins=30, density=True)
plt.title("Histogram of Customer Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.ylabel("Density")
plt.show()
```

```python
# Estimate mean and standard deviation
mu, sigma = stats.norm.fit(purchase_amounts)

print("Estimated Mean (μ):", mu)
print("Estimated Standard Deviation (σ):", sigma)
```

```python
plt.figure()

# Histogram of observed data
plt.hist(purchase_amounts, bins=30, density=True)

# Normal distribution curve
x = np.linspace(min(purchase_amounts), max(purchase_amounts), 200)
pdf = stats.norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title("Histogram with Fitted Normal Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Density")
plt.show()
```

```python
threshold = 120

# P(X > threshold)
probability = 1 - stats.norm.cdf(threshold, mu, sigma)

print(f"Probability that a customer spends more than ${threshold}: {probability:.4f}")
```

```python
simulated_sales = np.random.normal(mu, sigma, size=len(purchase_amounts))
```

```python
plt.figure()

plt.hist(purchase_amounts, bins=30, density=True, alpha=0.6, label="Observed Data")
plt.hist(simulated_sales, bins=30, density=True, alpha=0.6, label="Simulated Data")

plt.title("Observed vs Simulated Sales Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
print("Summary:")
print(f"Mean purchase amount: {mu:.2f}")
print(f"Standard deviation: {sigma:.2f}")
print(f"Probability of spending above ${threshold}: {probability:.2%}")
```
