```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Reproducibility
np.random.seed(42)

# Simulated customer purchase amounts (â‚¹)
purchase_amounts = np.random.normal(loc=1500, scale=300, size=500)

df = pd.DataFrame({
    "purchase_amount": purchase_amounts
})

df.head()
```

```python
# Calculate mean and standard deviation
mean = df["purchase_amount"].mean()
std = df["purchase_amount"].std()
```

```python
# Histogram
plt.figure()
plt.hist(df["purchase_amount"], bins=30, density=True)

# Normal distribution curve
x = np.linspace(df["purchase_amount"].min(), df["purchase_amount"].max(), 100)
pdf = norm.pdf(x, mean, std)
```

```python
plt.plot(x, pdf)
plt.xlabel("Purchase Amount")
plt.ylabel("Density")
plt.title("Histogram of Customer Purchase Amounts with Normal Fit")
plt.show()
```

```python
threshold = 1500  # example threshold

# Probability using normal distribution
probability_above_threshold = 1 - norm.cdf(threshold, mean, std)

probability_above_threshold
```

```python
# Simulated theoretical sales data using same mean & std
simulated_sales = np.random.normal(mean, std, 1000)

plt.figure()
plt.hist(df["purchase_amount"], bins=30, density=True, alpha=0.6, label="Actual Data")
plt.hist(simulated_sales, bins=30, density=True, alpha=0.6, label="Simulated Data")
```

```python
plt.xlabel("Purchase Amount")
plt.ylabel("Density")
plt.title("Actual vs Simulated Sales Distribution")
plt.legend()
plt.show()
```
