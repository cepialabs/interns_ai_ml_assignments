1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Load the Dataset

```python
df = pd.read_csv("Customer Purchase.csv")
df
```

3. Basic Data Exploration

```python
df.info()
```

```python
df.describe()
```

4. Purchase Count Analysis

```python
df['Purchased'].value_counts()
```

5. Purchase Distribution Plot

```python
df['Purchased'].value_counts().plot(kind='bar')
plt.title("Purchase Distribution")
plt.xlabel("Purchased")
plt.ylabel("Number of Customers")
plt.show()
```

6. Gender vs Purchase Analysis

```python
gender_purchase = pd.crosstab(df['Gender'], df['Purchased'])
gender_purchase
```

```python
gender_purchase.plot(kind='bar')
plt.title("Gender vs Purchase")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

7. Average Age of Purchased vs Not Purchased

```python
df.groupby('Purchased')['Age'].mean()
```

8. Probability of Purchase

```python
total_customers = len(df)
purchased_customers = len(df[df['Purchased'] == 'Yes'])

purchase_probability = purchased_customers / total_customers
purchase_probability
```

Interpretation:
This value represents the probability that a randomly selected customer makes a purchase.

9. Review Impact on Purchase

```python
review_purchase = pd.crosstab(df['Review'], df['Purchased'])
review_purchase
```

```python
review_purchase.plot(kind='bar')
plt.title("Review vs Purchase")
plt.xlabel("Review")
plt.ylabel("Count")
plt.show()
```

Conclusion

Customers with Good or Excellent reviews are more likely to purchase

Younger customers show slightly higher purchase behavior

Gender and education also influence purchase decisions

Probability analysis helps estimate customer conversion rate
