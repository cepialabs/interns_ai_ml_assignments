```python
import pandas as pd
import numpy as np
```

```python
# Probability of clicking the email
p_click = 0.30
p_no_click = 1 - p_click


# Probability of conversion given click
p_convert_given_click = 0.20
p_no_convert_given_click = 1 - p_convert_given_click


# Probability of conversion given no click
p_convert_given_no_click = 0.01
p_no_convert_given_no_click = 1 - p_convert_given_no_click
```

```python
click_cpt = pd.DataFrame({
"Click": ["Yes", "No"],
"Probability": [p_click, p_no_click]
})


print(click_cpt)
```

```python
conversion_cpt = pd.DataFrame({
"Click": ["Yes", "Yes", "No", "No"],
"Convert": ["Yes", "No", "Yes", "No"],
"Probability": [
p_convert_given_click,
p_no_convert_given_click,
p_convert_given_no_click,
p_no_convert_given_no_click
]
})


print(conversion_cpt)
```

```python
joint_probabilities = pd.DataFrame({
"Click": ["Yes", "Yes", "No", "No"],
"Convert": ["Yes", "No", "Yes", "No"],
"Joint Probability": [
p_click * p_convert_given_click,
p_click * p_no_convert_given_click,
p_no_click * p_convert_given_no_click,
p_no_click * p_no_convert_given_no_click
]
})


print(joint_probabilities)
```

```python
overall_conversion_rate = (
p_click * p_convert_given_click +
p_no_click * p_convert_given_no_click
)


print(f"Overall Conversion Rate: {overall_conversion_rate:.2%}")
```

```python
funnel = pd.DataFrame({
"Stage": ["Email Sent", "Clicked", "Converted"],
"Probability": [1.0, p_click, overall_conversion_rate]
})


print(funnel)
```
