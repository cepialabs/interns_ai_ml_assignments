## Marketing campaign Data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("marketing_campaign.csv", sep=';')
df.head()

```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

Calculate avg spending per customer

```python
spend_cols = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]

df['TotalSpend'] = df[spend_cols].sum(axis=1)
df[['ID', 'TotalSpend']].head()


```

```python
avg_spend = df[['ID', 'TotalSpend']]
avg_spend.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(avg_spend['TotalSpend'], kde=True)
plt.title("Distribution of Total Customer Spending")
plt.show()
```

```python
threshold = avg_spend['TotalSpend'].quantile(0.90)
threshold
```

```python
top_10_percent = avg_spend[avg_spend['TotalSpend'] >= threshold]
top_10_percent.head()
```

Visualization on top 10%

```python
df['SpenderCategory'] = np.where(
    df['TotalSpend'] >= threshold,
    'Top 10%',
    'Others'
)

sns.countplot(x='SpenderCategory', data=df)
plt.title("Top 10% Spenders vs Others")
plt.show()
```

Calculating skew

```python
engagement_cols = [
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth'
]

df[engagement_cols].skew()
```

Visualise engagement distribution

```python
for col in engagement_cols:
    plt.figure(figsize=(5,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```
