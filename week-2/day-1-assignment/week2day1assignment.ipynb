```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Marketing_Campaign_Data.xlsx")
df.head()
```

```python
average_spending = df['Cost (₹)'].mean()
print("Average spending per customer:", average_spending)
```

```python
threshold = df['Cost (₹)'].quantile(0.90)
print("Top 10% spending threshold:", threshold)
```

```python
top_10_spenders = df[df['Cost (₹)'] >= threshold]
top_10_spenders[['CampaignID', 'Cost (₹)']]
```

```python
clicks_skew = df['Clicks'].skew()
print("Clicks skewness:", clicks_skew)
```

```python
impressions_skew = df['Impressions'].skew()
print("Impressions skewness:", impressions_skew)
```

```python
leads_skew = df['Leads'].skew()
print("Leads skewness:", leads_skew)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Clicks'], bins=30, kde=True)
plt.title("Clicks Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Impressions'], bins=30, kde=True)
plt.title("Impressions Distribution")
plt.show()
```
