```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
# Load the dataset
df = pd.read_csv("customer_time_series (1).csv")

# Display first few rows
df.head()
```

```python
# Calculate customer retention
retention_df = (
    df.groupby(['Customer_ID', 'Account_Type'])
      .size()
      .reset_index(name='return_count')
)

retention_df.head()
```

```python
# Separate account types
current_accounts = retention_df[
    retention_df['Account_Type'] == 'Current'
]['return_count']

savings_accounts = retention_df[
    retention_df['Account_Type'] == 'Savings'
]['return_count']

print("Current Account count:", len(current_accounts))
print("Savings Account count:", len(savings_accounts))
```

```python
# Perform independent t-test
t_stat, p_value = ttest_ind(
    current_accounts,
    savings_accounts,
    equal_var=False
)

t_stat, p_value
```

```python
# Perform independent t-test
t_stat, p_value = ttest_ind(
    current_accounts,
    savings_accounts,
    equal_var=False
)

t_stat, p_value
```

### Hypothesis

- **Null Hypothesis (H₀):**
  There is no difference in customer retention between account types.

- **Alternative Hypothesis (H₁):**
  There is a significant difference in customer retention between account types.


### Interpretation

- If p-value < 0.05:
  There is a statistically significant difference in customer retention.

- If p-value ≥ 0.05:
  There is no statistically significant difference in customer retention.


### Conclusion (Plain English)

A t-test was conducted to examine whether customer retention differs
between Current and Savings account holders.

Customer retention was measured by counting how many times a customer
appeared in the dataset, which represents how often they returned.

Based on the obtained p-value, we conclude that the difference in
customer retention between account types is
(statistically significant / not statistically significant).

This means that account type
(does / does not) have a meaningful impact on customer retention.
