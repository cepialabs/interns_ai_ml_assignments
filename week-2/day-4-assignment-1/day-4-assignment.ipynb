```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("advertising.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Missing values per column:\n")
df.isnull().sum()
```

```python
# Fill missing numeric values with mean
df = df.fillna(df.mean(numeric_only=True))
# Remove duplicate rows
df = df.drop_duplicates()
print("After cleaning:")
df.info()
```

```python
#Correlation Analysis 
corr_matrix = df.corr(numeric_only=True)
print("Correlation with Sales:\n")
print(corr_matrix["Sales"])
```

```python
#Visualize Relationships (Scatter Plots)
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.scatter(df["TV"], df["Sales"])
plt.xlabel("TV Spend")
plt.ylabel("Sales")
plt.title("TV vs Sales")

plt.subplot(1,3,2)
plt.scatter(df["Radio"], df["Sales"])
plt.xlabel("Radio Spend")
plt.title("Radio vs Sales")

plt.subplot(1,3,3)
plt.scatter(df["Newspaper"], df["Sales"])
plt.xlabel("Newspaper Spend")
plt.title("Newspaper vs Sales")

plt.tight_layout()
plt.show()
```

```python
#Business Insight (Feature Importance)
print("Feature importance based on correlation:")
print(corr_matrix["Sales"].sort_values(ascending=False))
```

```python
#Save cleaned dataset
df.to_csv("cleaned_marketing_sales.csv", index=False)
print("Cleaned dataset saved as cleaned_marketing_sales.csv")
```
