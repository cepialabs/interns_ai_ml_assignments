```python
#Calculate Average Spending per Customer
# Step 1: Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Print columns to verify
print("Available columns:", df.columns.tolist())

# Step 3: Automatically find spending column
spend_column = [col for col in df.columns if 'spend' in col or 'amount' in col or 'cost' in col][0]

print("Using spending column:", spend_column)

# Step 4: Calculate Average Spending per Customer
average_spend = df[spend_column].mean()
print("Average Spending per Customer:", average_spend)
```

```python
#Identify Top 10% of Spenders
# Step 1: Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Automatically detect spending column
spend_column = [col for col in df.columns if 'spend' in col or 'amount' in col or 'cost' in col][0]

# Step 3: Calculate 90th percentile (Top 10% threshold)
threshold = df[spend_column].quantile(0.90)

# Step 4: Filter Top 10% Spenders
top_10_spenders = df[df[spend_column] >= threshold]

# Step 5: Display results
print("Spending Column Used:", spend_column)
print("Top 10% Spending Threshold:", threshold)
print("Top 10% Spenders:")
top_10_spenders.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Step 1: List all numeric columns in dataset
numeric_cols = df.select_dtypes(include='number').columns.tolist()
print("Numeric columns detected:", numeric_cols)

# Step 2: Remove ID or irrelevant numeric columns (like 'ID', 'Year_Birth') if needed
plot_cols = [col for col in numeric_cols if col not in ['ID', 'Year_Birth']]

print("Columns we will plot:", plot_cols)

```

```python
# Step 3: SCATTER PLOT (first two numeric columns)
if len(plot_cols) >= 2:
    plt.figure(figsize=(6,4))
    plt.scatter(df[plot_cols[0]], df[plot_cols[1]])
    plt.title(f"{plot_cols[0]} vs {plot_cols[1]}")
    plt.xlabel(plot_cols[0])
    plt.ylabel(plot_cols[1])
    plt.show()
```

```python
# Step 4: PIE CHART (Proportion of total per column)
total_vals = df[plot_cols].sum()
plt.figure(figsize=(6,6))
plt.pie(total_vals.values, labels=total_vals.index, autopct='%1.1f%%')
plt.title("Proportion of Numeric Columns")
plt.show()
```

```python
# Step 5: BAR PLOT (Average of each numeric column)
avg_vals = df[plot_cols].mean()
plt.figure(figsize=(8,4))
plt.bar(avg_vals.index, avg_vals.values)
plt.title("Average Values of Numeric Columns")
plt.xlabel("Columns")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 6: HISTOGRAMS
for col in plot_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
