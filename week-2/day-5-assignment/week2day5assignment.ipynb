```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python

df = pd.read_csv("shopping_trends.csv")
```

```python
df["Discount Applied"] = df["Discount Applied"].str.strip()
```

```python
discount_yes = df[df["Discount Applied"] == "Yes"]["Previous Purchases"]
discount_no  = df[df["Discount Applied"] == "No"]["Previous Purchases"]
```

```python
print("Mean Previous Purchases (Discount Applied):", discount_yes.mean())
print("Mean Previous Purchases (No Discount):", discount_no.mean())
```

```python
t_stat, p_value = ttest_ind(
    discount_yes,
    discount_no,
    equal_var=False
)
```

```python
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
alpha = 0.05
```

```python
if p_value <= alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
