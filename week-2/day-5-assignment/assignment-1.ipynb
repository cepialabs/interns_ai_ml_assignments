```python
#import libraries

import pandas as pd
from scipy.stats import ttest_ind
```

```python
#load dataset

df = pd.read_csv("02 Churn-Dataset.csv")
df.head()
```

```python
#Basic Data Check

df.columns
```

```python
df.isnull().sum()
```

```python
# Convert TotalCharges to numeric 
if 'TotalCharges' in df.columns:
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Drop rows with missing values
df = df.dropna(subset=['Churn', 'MonthlyCharges'])
```

```python
#Split Data into Two Groups

returned = df[df['Churn'] == 'No']['MonthlyCharges']
not_returned = df[df['Churn'] == 'Yes']['MonthlyCharges']

print("Returned customers count:", returned.shape[0])
print("Not returned customers count:", not_returned.shape[0])
```

```python
#Perform Independent Two-Sample t-Test

t_stat, p_value = ttest_ind(returned, not_returned, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Result: Statistically significant difference between returned and not returned customers.")
else:
    print("Result: No statistically significant difference between returned and not returned customers.")
```

An independent two-sample t-test was conducted to compare the monthly charges of customers who returned and those who did not return.
The results showed a statistically significant difference (p < 0.05), indicating that monthly charges vary meaningfully between the two groups.
This suggests that pricing may be associated with customer retention behavior rather than occurring by random chance.

(If p ≥ 0.05, replace “statistically significant” with “not statistically significant.”)
