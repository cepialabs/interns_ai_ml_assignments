```python
#Split into two groups

retained = df[df["retained"] == 1]["avgorder"].values
not_retained = df[df["retained"] == 0]["avgorder"].values
```

```python
#Manual t-test calculation

n1 = len(retained)
n2 = len(not_retained)

mean1 = np.mean(retained)
mean2 = np.mean(not_retained)

var1 = np.var(retained, ddof=1)
var2 = np.var(not_retained, ddof=1)

t_stat = (mean1 - mean2) / np.sqrt((var1/n1) + (var2/n2))
print("T-statistic:", t_stat)
```

```python
#P-value

from math import erf, sqrt

def normal_cdf(x):
    return (1 + erf(x / sqrt(2))) / 2

p_value = 2 * (1 - normal_cdf(abs(t_stat)))
print("P-value:", p_value)
```

```python
#Final decision

alpha = 0.05

if p_value < alpha:
    print("Result: Retention has significant effect on avg order")
else:
    print("Result: No significant relationship")
```

```python
#Report

We conducted a manual independent t-test to compare the average order value between retained and non-retained customers.
The t-statistic was calculated using mathematical formula without any statistical library.
The obtained p-value shows whether the difference is statistically significant.
If p-value is less than 0.05, we conclude that retention significantly affects customer spending behavior.
```
