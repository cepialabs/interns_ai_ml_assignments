```python
#Correlation
x = df['marketing_budget(thousands)']
y = df['actual_sales(millions)']

correlation = x.corr(y)
print("Correlation:", correlation)
```

```python
#Manual Pearson Formula
numerator = ((x - x.mean()) * (y - y.mean())).sum()
denominator = np.sqrt(((x - x.mean())**2).sum() * ((y - y.mean())**2).sum())

pearson_corr = numerator / denominator
pearson_corr
```

```python
# scatter plot
plt.figure()
plt.scatter(x, y)
plt.xlabel("Marketing Budget (thousands)")
plt.ylabel("Actual Sales (millions)")
plt.title("Marketing Budget vs Sales")
plt.show()
```

```python
#Regression Line

m, b = np.polyfit(x, y, 1)

plt.figure()
plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.xlabel("Marketing Budget")
plt.ylabel("Sales")
plt.title("Regression Line")
plt.show()
```

```python
#Prediction

new_budget = 50   # 50 thousand
predicted_sales = m * new_budget + b
predicted_sales
```
