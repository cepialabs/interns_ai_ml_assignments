```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
X = np.array([
    [2, 3],
    [1, 5],
    [2, 1],
    [3, 2]
])

y = np.array([
    [1],
    [1],
    [0],
    [1]
])
```

```python
input_size = 2
hidden_size = 3
output_size = 1
```

```python
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
z1 = np.dot(X, W1) + b1
a1 = sigmoid(z1)

z2 = np.dot(a1, W2) + b2
y_pred = sigmoid(z2)
```

```python
loss = np.mean((y - y_pred) ** 2)
```

```python
d_output = (y_pred - y) * sigmoid_derivative(y_pred)
```

```python
dW2 = np.dot(a1.T, d_output)
db2 = np.sum(d_output, axis=0, keepdims=True)
```

```python
d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(a1)
```

```python
dW1 = np.dot(X.T, d_hidden)
db1 = np.sum(d_hidden, axis=0, keepdims=True)
```

```python
learning_rate = 0.1

W1 -= learning_rate * dW1
b1 -= learning_rate * db1

W2 -= learning_rate * dW2
b2 -= learning_rate * db2
```

```python
learning_rate = 0.1

for epoch in range(1000):

    # Forward propagation
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)

    # Loss
    loss = np.mean((y - y_pred) ** 2)

    # Backpropagation
    d_output = (y_pred - y) * sigmoid_derivative(y_pred)

    dW2 = np.dot(a1.T, d_output)
    db2 = np.sum(d_output, axis=0, keepdims=True)

    d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(a1)

    dW1 = np.dot(X.T, d_hidden)
    db1 = np.sum(d_hidden, axis=0, keepdims=True)

    # Update weights
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
test = np.array([[3,3]])

z1 = np.dot(test, W1) + b1
a1 = sigmoid(z1)

z2 = np.dot(a1, W2) + b2
prediction = sigmoid(z2)

print("Pass probability:", prediction)
```
