Import all required libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Create Dataset

```python
# Input data (Years of Experience)
X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])

# Output labels (0 = No Promotion, 1 = Promotion)
y = np.array([0,0,0,0,1,1,1,1])

print("Input Data:")
print(X)

print("Output Labels:")
print(y)
```

Build Neural Network Model

```python
model = Sequential()

# Hidden layer
model.add(Dense(5, activation='relu', input_dim=1))

# Output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

Compile the Model

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

Train the model

```python
history = model.fit(X, y, epochs=100)
```

Make Prediction

```python
prediction = model.predict(np.array([[5]]))

print("Prediction value:", prediction)
```

```python
if prediction > 0.5:
    print("Predicted Result: Promotion")
else:
    print("Predicted Result: No Promotion")
```
