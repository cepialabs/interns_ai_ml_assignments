```python
import torch
import torch.nn as nn
x = torch.tensor([[1.0, 2.0, 3.0]])  # Input data
y = torch.tensor([[10.0]])           # The correct answer we want to predict

model = nn.Sequential(
    # Input layer (3 features) to Hidden layer (4 neurons)
    nn.Linear(3, 4),
    # Activation function
    nn.ReLU(),  
    # Hidden layer (4 neurons) to Output layer (1 prediction)
    nn.Linear(4, 1)   
)
# The tool that calculates the error
loss_fn = nn.MSELoss() 
# The tool that updates weights
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) 

print("Starting Training...")

for epoch in range(10):
    # STEP 1: Forward Pass (Input to Output)
    prediction = model(x)
    
    # STEP 2: Calculate the Loss 
    loss = loss_fn(prediction, y)
    
    # STEP 3: Backward Pass (Calculate the gradients)
    loss.backward()
    
    # STEP 4: Update Weights 
    optimizer.step()
    
    # Clear out old calculations for the next loop
    optimizer.zero_grad() 
    
    print(f"Loop {epoch+1} | Prediction: {prediction.item():.2f} | Loss: {loss.item():.2f}")
```
