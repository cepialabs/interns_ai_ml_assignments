```python
import torch
import torch.nn as nn
```

```python
x = torch.tensor([[1.0, 2.0, 3.0]]) 
y = torch.tensor([[10.0]])    
```

```python
model = nn.Sequential(
    # Input layer to Hidden layer 
    nn.Linear(3, 4),
    nn.ReLU(),  
    nn.Linear(4, 1)   
)
loss_fn = nn.MSELoss() 
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) 
```

```python
print("Starting Training...")
for epoch in range(10):
    prediction = model(x)
    loss = loss_fn(prediction, y)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad() 
    
    print(f"Loop {epoch+1} | Prediction: {prediction.item():.2f} | Loss: {loss.item():.2f}")
```
