### What is a neural network?
- A neural network is a machine learning model inspired by the human brain.
- It consists of layers of neurons that process input data and learn patterns.
- A basic neural network has three main layers: input layer, hidden layer, and output layer.
- During training, data flows through these layers and the model updates its weights
- using backpropagation to reduce error.

### The layers of the neural network is:
- Input Layer:
    - Receives the input data.

- Hidden Layer:
    - Processes the input using weights, bias, and activation functions.

- Output Layer:
    - Produces the final prediction or result.

#### Simple Neural Network Code with one hidden layer:

```python
import numpy as np
```

```python
# Input data
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
```

```python
# Output
y = np.array([[0],[1],[1],[0]])
```

```python
# Sigmoid activation
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Weights initialization
w1 = np.random.rand(2,3)
w2 = np.random.rand(3,1)
```

```python
# Forward propagation
hidden = sigmoid(np.dot(X, w1))
output = sigmoid(np.dot(hidden, w2))

print(output)
```

### Data Flow During Training

Step 1: Input Layer  
The input data enters the neural network.

Example:
X = [1,0]

Step 2: Hidden Layer  
The input values are multiplied with weights and bias is added.

Formula:
Z = XW + b

An activation function (sigmoid) is applied to introduce non-linearity.

Step 3: Output Layer  
The hidden layer output is passed to the output layer where another
weighted sum and activation function produce the final prediction.

Step 4: Error Calculation  
The predicted output is compared with the actual output.

Error = Actual Output - Predicted Output

Step 5: Backpropagation  
The error is propagated backward and weights are updated to reduce
the prediction error.

This process repeats for multiple iterations until the model learns
the correct pattern.
