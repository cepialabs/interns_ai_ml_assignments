```python
#Cell 1 вҖ” Import Libraries

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Cell 2 вҖ” Create Dataset
# Input data (hours studied)
X = np.array([[1],[2],[3],[4],[5]])

# Output labels (0 = Fail, 1 = Pass)
y = np.array([0,0,0,1,1])

print("Input Data:")
print(X)

print("Output Labels:")
print(y)
```

```python
#Cell 3 вҖ” Build Neural Network Model
# Create model
model = Sequential()

# Hidden layer with 5 neurons
model.add(Dense(5, activation='relu', input_dim=1))

# Output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
#Cell 4 вҖ” View Model Structure
model.summary()
```

```python
#Cell 5 вҖ” Compile the Model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
#Cell 6 вҖ” Train the Model
history = model.fit(X, y, epochs=100)
```

```python
#Cell 7 вҖ” Make Prediction
prediction = model.predict(np.array([[3]]))

print("Prediction value:", prediction)
```

```python
#Cell 8 вҖ” Interpret Prediction
if prediction > 0.5:
    print("Predicted Result: Pass")
else:
    print("Predicted Result: Fail")
```

```python
#Cell 9 вҖ” Explanation of Data Flow
print("""
Data Flow in Neural Network:

1. Input layer receives study hours as input.
2. Data moves to the hidden layer where weights and activation functions process the input.
3. Forward propagation generates a prediction.
4. Loss function calculates the error between predicted and actual output.
5. Backpropagation adjusts weights to reduce the error.
6. After multiple epochs, the model improves prediction accuracy.
""")
```
