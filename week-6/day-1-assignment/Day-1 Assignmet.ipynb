## Problem Statement

### Student Pass Prediction

Build a simple neural network to predict whether a student **passes or fails** based on the following features:

- **Study Hours**
- **Sleep Hours**

### Output

- **1 → Pass**
- **0 → Fail**
### Neural network will have:

- **2 → Input Neurons**

- **3 → Hidden Neurons**

- **1 → Output Neuron**

```python
# step-1 import libraries
import numpy as np
```

```python
#step-2 imput data
x= np.array([
    [1,2],
    [2,3],
    [3,5],
    [5,6]
])
```

```python
#Step-3 actual output 
y=np.array([[0],[0],[1],[1]])
```

```python
#Step-4  weights initialization
weights_input_hidden = np.random.rand(2,3)
weights_hidden_output=np.random.rand(3,1)
```

```python
#Step-5 learning rate
lr = 0.1
```

```python
#Step-6 activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#Step-7 derivative of sigmoid
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
#Step-8 training loop
for epoch in range(10000):
      # ----- Forward propagation -----
    hidden_layer = sigmoid(np.dot(x, weights_input_hidden))
    output_layer = sigmoid(np.dot(hidden_layer, weights_hidden_output))

    # ----- Error calculation -----
    error = y - output_layer

    # ----- Backpropagation -----

    # output layer gradient
    d_output = error * sigmoid_derivative(output_layer)

    # hidden layer error
    error_hidden = d_output.dot(weights_hidden_output.T)

    # hidden layer gradient
    d_hidden = error_hidden * sigmoid_derivative(hidden_layer)

    # ----- Update weights -----
    weights_hidden_output += hidden_layer.T.dot(d_output) * lr
    weights_input_hidden += x.T.dot(d_hidden) * lr
```

```python
print("Trained Output:")
print(output_layer)
```

```python
# Step-8 Convert Probabilities to Pass / Fail
print("\nPass/Fail Prediction:")
for i in range(len(output_layer)):
    result = "Pass" if output_layer[i][0] >= 0.5 else "Fail"
    print(f"Input: {x[i]}  Output: {output_layer[i][0]:.3f}  Result: {result}")
```

## Complete Data Flow of Neural Network

### Step-by-Step Training Process

1. **Input Data** – The neural network receives input features such as Study Hours and Sleep Hours.

2. **Multiply with Weights** – Each input value is multiplied by a weight.

3. **Activation Function** – The activation function introduces non-linearity.

4. **Hidden Layer Processing** – The hidden layer processes the transformed inputs.

5. **Output Prediction** – The network predicts the output (Pass or Fail).

6. **Loss Calculation** – The difference between predicted output and actual output is calculated.

7. **Backpropagation** – The error is propagated back through the network.

8. **Weight Update** – Weights are adjusted to reduce prediction error.

9. **Repeat Training** – The process repeats until the model learns the correct pattern.
