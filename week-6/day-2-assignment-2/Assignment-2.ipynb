```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
# Load dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Flatten images
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
```

```python
# ------------------------------
# Model 1 : ReLU Activation
# ------------------------------
model_relu = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_relu.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])

history_relu = model_relu.fit(x_train, y_train,
                              epochs=5,
                              validation_data=(x_test, y_test))
```

```python
# ------------------------------
# Model 2 : Sigmoid Activation
# ------------------------------
model_sigmoid = keras.Sequential([
    layers.Dense(128, activation='sigmoid', input_shape=(784,)),
    layers.Dense(64, activation='sigmoid'),
    layers.Dense(10, activation='softmax')
])

model_sigmoid.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

history_sigmoid = model_sigmoid.fit(x_train, y_train,
                                    epochs=5,
                                    validation_data=(x_test, y_test))
```

```python
# Compare Performance
print("ReLU Model Accuracy:", history_relu.history['accuracy'][-1])
print("Sigmoid Model Accuracy:", history_sigmoid.history['accuracy'][-1])
```

```python
# Plot Comparison Graph
plt.plot(history_relu.history['accuracy'], label='ReLU')
plt.plot(history_sigmoid.history['accuracy'], label='Sigmoid')

plt.title("Training Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
# Expected Result (Observation)
| Activation | Training Speed | Accuracy | Problem                     |
| ---------- | -------------- | -------- | --------------------------- |
| ReLU       | Faster         | Higher   | Works well in deep networks |
| Sigmoid    | Slower         | Lower    | Vanishing gradient problem  |
```
